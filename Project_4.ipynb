# 1. Nonlinear Susceptibilities Calculated Using Time-Independent Perturbation Theory

<p style='text-align:justify;'>

We will use time-independent perturbation theory as it is easier to implement as compared to time-dependent perturbation theory.
The goal is to study how materials respond to intense laser light when the laser's frequency $\omega $ is much higher than the natural frequency $\omega_0 $ of the material as we want to avoid absorption losses.

Let's dive into the mathematics and do a derivation of how to calculate the nonlinear susceptibilities $\chi^{(n)}$ using time-independent perturbation theory. We'll start with the formula for the energy stored in a material due to its polarization:


$$W = -\int  \tilde P(\tilde{E '}) d\tilde{E'} $$

Here, $\tilde P({\tilde E'})$ is polarization of the material at a particular value of ${\tilde E}$.
Let's express the polarization in terms of its susceptibility terms:

$$\tilde{P} = \varepsilon_0 \left(\chi^{(1)} \tilde{E} + \chi^{(2)} \tilde{E}^2 + \chi^{(3)}\tilde{E}^3 + \ldots\right)$$

Now, we can substitute this expression into the energy equation:

$$W = -\varepsilon_0 \int \left(\chi^{(1)} \tilde{E}^2 + \chi^{(2)} \tilde{E}^3 + \chi^{(3)} \tilde{E}^4 + \ldots\right) d\tilde{E}$$
$$W = -\int \tilde P(\tilde{E} )\, d\tilde{E} = -\varepsilon_0\frac{1}{2} \chi^{(1)} E^2 - \varepsilon_0\frac{1}{3} \chi^{(2)} E^3 - \varepsilon_0\frac{1}{4} \chi^{(3)} E^4 - \ldots \equiv W^{(2)} + W^{(3)} + W^{(4)} + \ldots$$

    
We know $ W $ as a power series
in $\tilde{E}$ we can determine the susceptibilities a
    
$$\chi^{(n-1)} = -n\frac{ W^{(n)}}{\epsilon_0 E^{(n)}}$$

Also, the nonlinear susceptibilities
can be obtained through differentiation as

$$\chi^{(n-1)} = -\frac{1}{{\varepsilon_0(n - 1)!}} \frac{{\partial^nW}}{{\partial\tilde{E}^n}}\bigg|_{E=0}$$
.

# General Expression for the Nonlinear Susceptibility in the Quasi-Static Limit

<p style='text-align:justify;'>

Problem - In quantum mechanics involves determining how the energy of some state  $|\psi_n\rangle$
 of an atomic system is modified in response to a perturbation of the atom.
    
Solution- We assume tha the Hamiltonian ($\hat{H}$) of the atomic system, which is represented as the sum of two terms:
$\hat{H}_0$, which represents the total energy of the free atom, and $\hat{V}$, which represents the quasi-static perturbation due to an external electric field.

   $$\hat{H} = \hat{H}_0 + \hat{V}$$


The perturbation term $\hat{V}$ is specifically related to an applied quasi-static electric field ($\tilde{E}$).
   - The electric dipole moment operator ($\hat{\mu}$) is represented as $\hat{\mu} = -e\hat{x}$.

   $$\hat{V} = -\hat{\mu} \tilde{E}$$

The atomic system's wavefunction is described by the time-independent Schrödinger equation.
   
   $$\hat{H} |\psi_n\rangle = w_n |\psi_n\rangle$$

   Where $|\psi_n\rangle$ represents a state of the system, and $w_n$ is the energy of that state.

To solve this equation, energy ($w_n$) and the state vector ($|\psi_n\rangle$) are represented as power series in terms of the perturbation.

   $$w_n = w^ {(0)}_n + w^{(1)}_n + w^{(2)}_n + \ldots$$
   $$|\psi_n\rangle = |\psi^{(0)}_n\rangle + |\psi^{(1)}_n\rangle + |\psi^{(2)}_n\rangle + \ldots$$

The expressions for the first four orders of energy corrections due to the perturbation:
   
   $$w^{(1)}_n = e\tilde{E}\langle n|x|n\rangle$$
    
   $$w^{(2)}_n = e^2\tilde{E}^2 \sum_s \frac{ \langle n|{x}|s\rangle \langle s|\hat{x}|n\rangle}{ {(w^{(0)}_s - w^{(0)}_n)}{(w^{(0)}_s - w^{(0)}_n)}}$$
    
   $$w^{(3)}_n = e^3\tilde{E}^3 \sum_{s,t} \frac {\langle n|x|s\rangle \langle s|x|t\rangle \langle t|x|n\rangle} {(w^{(0)}_s - w^{(0)}_n)(w^{(0)}_t - w^{(0)}_n)}$$
    
   $$w^{(4)}_n = e^4\tilde{E}^4 \sum_{s,t,u} \frac {\langle n|x|s\rangle \langle s|x|t\rangle \langle t|{x|u}\rangle \langle u|x|n\rangle}{ (w^{(0)}_s - w^{(0)}_n)(w^{(0)}_t - w^{(0)}_n)(w^{(0)}_u - w^{(0)}_n)} - e^2\tilde{E}^2 w^{(2)}_n \sum_u \frac{\langle n|x|u\rangle }{(w^{(0)}_u - w^{(0)}_n)^2 } $$

where ${\hbar w_{sg} = w^{(0)}_{s}-w^{(0)}_{s}}$ and so on. We see that $\chi^{(3)}$ naturally decomposes into the sum
of two terms, which can be represented schematically in terms of the two diagrams shown in following figure

Screenshot (19).png

To simplify these expressions we can use the Unsöld approximation, which entails replacing each resonance
frequency (e.g., $w_{sg}$) by some average transition frequency $ω_{0}$.
    
$$\alpha = \frac{2e^2}{\hbar \omega_0} \sum_s \langle g | x | s \rangle \langle s | x | g \rangle
$$

$\alpha = \frac{2e^2}{\hbar \omega_0} \langle g | x \hat{O} x | g \rangle
$ where  ${\hat{O}= \sum_s {\left|s\right\rangle\left\langle s\right|}}$.

We thus find that
$$\alpha = -\frac{2e^2}{{\hbar}\omega_0^2} \langle x^2 \rangle
$$    
This result shows that the linear susceptibility is proportional to the electric quadrupole moment
of the ground-state electron distribution.
    
For the second- and third-order nonlinear coefficients to find that   
$$\beta = -\frac{3e^3}{{\hbar}^2\omega_0^2} \langle x^3 \rangle
$$
$$\gamma = \frac{4e^4}{{\hbar}^3\omega_0^3}[ \langle x^4 \rangle - 2\langle x^2 \rangle^2]
$$


This suggest that the linear and nonlinear polarizabilities increase rapidly with the physical dimensions of the electron cloud associated with the atomic ground state.

The hyperpolarizabilities ($\beta$ and $\gamma$) are interpreted as measures of higher-order moments of the ground-state electron distribution (e.g., octopole and hexadecapole moments).
   - $\gamma$ is expressed in terms of a dimensionless quantity 'g' known as kurtosis, which provides a measure of the normalized fourth moment of the electron distribution.

$$\gamma = \alpha^2 \frac{ g} {\hbar \omega_0}
$$

$$g = [\frac{\langle x^4\rangle}{ \langle x^2 \rangle} - 2]
$$


The average transition frequency $\omega_0 $ is related to the moments of the electron distribution and is used to simplify the expressions for polarizabilities.We start with the Thomas–Reiche–Kuhn sum rule which states that
$$ \frac {2m}{\hbar}\sum_k \omega_{kg}|x_k g|^2 = Z
$$  
    
where $Z$ is the number of optically active electrons. If we now replace ωkg by the average transition frequency $ω_0$ and perform the summation over k in the same manner as in the derivation, we obtain
 $$\omega_0 =\frac{ Z \hbar}{2m{\langle x^2 \rangle}}$$

Now we can simplify,
  
$$\alpha = \frac{4e^2m}{ Z^2 {\hbar}^2} \langle x^2 \rangle^2$$
$$\beta = \frac {-12e^3m^2} { Z^2 {\hbar}^4} \langle x^2 \rangle^2 \langle x^3 \rangle$$
$$\gamma = \frac {32e^4m^3} { Z^3 {\hbar}^6} (\langle x^2 \rangle^3 \langle x^4 \rangle - 2 \langle x^2 \rangle^2)$$
    
Note  that $\alpha$ is a measure of the electric quadrupole moment of the ground-state electron distribution, $\beta$ is a measure of the octopole moment of the ground-state electron distribution, and
$\gamma$ depends on both the hexadecimal pole and the quadrupole moment of the electron ground state electron distribution.

    
The derived expressions provide a quantum-mechanical framework for understanding the linear and nonlinear responses of an atomic system to an external electric field, which are fundamental in the study of optical properties and interactions.


# **2. Semiempirical Models of the Nonlinear Optical Susceptibility.**
<p style='text-align:justify;'>

**Generalization of Miller's Rule:**

Miller's rule is a method used to predict second-order nonlinear optical properties.
The generalization of Miller's rule for third-order nonlinear optical interactions is presented as Equation.
The formula involves the product of first-order susceptibilities $\chi^{(1)}$ at different frequencies.
$$\chi^{(3)}(\omega_4, \omega_3, \omega_2, \omega_1) = A \chi^{(1)}(\omega_4) \chi^{(1)}(\omega_3) \chi^{(1)}(\omega_2) \chi^{(1)}(\omega_1)$$
where, $$\omega_4 = \omega_1 + \omega_2 + \omega_3$$
A is a frequency independent quantity that is nearly same for all material. This only holds for ionic crystals and breaksdown for materials such as low pressure gases.


**Wang's Relation:**

Wang’s relation is formulated for the nonlinear optical response in the quasi-static limit and states that
$$\chi^{(3)} = Q' \chi^{(1)^2}$$
where $$Q' = \frac{g'}{N_{\text{eff}} \omega_0}
$$

This relation involves a dimensionless parameter $g'$ is related to the kurtosis of the ground-state electron distribution. It is of the order of unity and assumed to be nearly same for all materials.$ N_{\text{eff}} = N_f$
where $N$ is molecular number density and $f$ is oscillator strength. $\omega_0$ is an average transition frequency.

This relation gives accurate results for both ionic crystals and low pressure gases.

**Model of Boling, Glass, and Owyoung:**

The model proposed by Boling et al. (1978) builds upon Wang's formula to predict the nonlinear refractive index constant $n_2$ based on linear optical properties. The linear refractive index $n$ is assumed to follow the Lorentz–Lorenz law and Lorentz oscillator model which are expressed by the following equations:
$$\frac{n_2 - 1}{n_2 + 2} = \frac{1}{3} N \alpha$$
$$\alpha = \frac{f e^2}{m(\omega_0^2 - \omega^2)}$$
where $4$ is the oscillator strength, $e$ is the elementary charge, $m$ is the mass, $ω_0$  is the resonance frequency.

By measuring the refractive index as a function of frequency, the resonance $ω_0$ and effective number density $Nf$ can be determined.

The Lorentz–Lorenz law describes the relationship between the refractive index and the density of polarizable particles in a medium. The Lorentz oscillator model, on the other hand, characterizes the response of a material to an electromagnetic field, incorporating a resonant frequency $ω_0$ and damping term $ω$. Both play key roles in understanding and predicting the linear optical properties of materials.
The nonlinear refractive index $n_2$ is determined using Equations:
$$\begin{align*}
n_2 &= \frac{3}{4n^2\epsilon_0c} \chi^{(3)} \\
\chi^{(3)} &= L^4 N \gamma \\
L &= \frac{n^2+2}{3} \\
\gamma &= g \alpha^2 \frac{1}{\hbar\omega_0}
\end{align*}
$$
Simplifying the expressions:
\begin{align*}n_2 &= \frac{1}{4\epsilon_0c} \left(\frac{n^2+2}{3}\right)^4 \frac{N}{n^2} (g \alpha^2 \frac{1}{\hbar\omega_0})
\end{align*}

which can also be written as:
$$n_2 = \frac{(n_2 + 2)^2 (n_2 - 1)^2 (gf)}{6n^2c\epsilon_0ħ \omega_0(Nf)}$$

The value $gf$ = 3 is found empirically to give good agreement with measured values.

The optical response of σ electrons in σ bonds differs significantly from that of delocalized π electrons in double bonds. π electrons are less tightly bound, allowing for more significant responses to an applied optical field.

# **5.5 Nonlinear Optics of Chiral Media**

In this, we will be discussing the optical properties of a chiral medium i.e. a medium composed of a collution of chiral molecules.

A chiral molecule is a molecule in which the mirror image looks different than the molecule itself. e.g.$CO_2$, $H_2O$,$CH_4$ are achiral.

chiral.jpeg


Many organic molecules are chiral.

It is well known that chiral media display the property of
optical activity, that is, the rotation of the direction of linear polarization of a light beam as it
passes through such a material.

IMPORTANT DEFINITIONS:

**1.Dextrorotatory material** :
The material in which the direction of polarization rotates in clockwise manner as beam propagates.


**2.Levorotatory material**:The material in which the direction of polarization rotates in clockwise manner as beam propagates.

**3.Enantiomers**: Two molecules which are mirror images of each other.

**4.Racemic mixture**:An equal mixture of two enantiomers is said to be a racemic mixture.

Now, let's discuss the non-linear optical properties of chiral molecules.

1.A liquid which is composed of chiral molecules is isotropic but noncentrosymmetric and thus it possess a second-order non-linear optical response.

2.Such a medium can produce sum- or difference-frequency generation, but not second-harmonic generation, and moreover can produce sum- or difference-frequency generation only if the two input beams are non-collinear.

Research Advancements:

The theory of second-order processes in chiral media was developed by Giordmaine (1965) and was studied experimentally by Rentzipis et al. (1966). More
recent research on the nonlinear optics of chiral media includes that of Verbiest et al. (1999)

Theoretical description of second-order processes in chiral materials:

We represent the second-order polarization induced in such a material as:

$P_i(\omega \sigma) = 2\varepsilon_0 \sum_{j,k} \chi^{(2)}_{ijk}(\omega \sigma = \omega_1 + \omega_2) E_j F_k$

where

$E_j$ represents a field at frequency $w_1$

$F_k$ represents a field at frequency $ω_2$

Now the final equation becomes;

$P_i = \varepsilon_0 \sum_{j,k} S_{ijk}(E_jF_k + E_kF_j) + \varepsilon_0 \sum_{j,k} A_{ijk}(E_jF_k - E_kF_j)$

where

$S_{ijk} = \frac{1}{2} \left(\chi^{(2)}_{ijk} + \chi^{(2)}_{ikj}\right)$


$A_{ijk} = \frac{1}{2} \left(\chi^{(2)}_{ijk} - \chi^{(2)}_{ikj}\right)$

For isotropic but noncentrosymmetric medium $S_{ijk}$ vanishes and the only nonvanishing elements of $A_{ijk}$ are

$A_{123} = A_{231} = A_{312}$
  
So the non linear polarization becomes:

$P = \varepsilon_0A_{123} \mathbf{E} \times \mathbf{F}$

The two input beams are at different frequencies, as required for $A_{123}$ to be nonzero. In addition, they are orthogonally polarized to ensure that E × F is nonzero and are noncollinear
to ensure that P has a transverse component. Generation of a sum-frequency signal at 2314 Å

#5.6 Nonlinear Optics of Liquid Crystals

Liquid crystal materials have large nonlinear optical effects. They are composed of large-anisotrophic molecules. Anisotropic molecules are molecules that have different properties or characteristics in different directions. In simple terms, these molecules aren't symmetric and they have a preferred orientation or alignment.

Above a certain transition temperature, which might typically be 100°C, these materials exist in an isotropic phase in which they behave like ordinary liquids.

In the isotropic phase the molecules in the liquid crystal material have a random or disordered orientation. They move around like molecules in a regular liquid, with no particular alignment. In this phase, the material behaves like an ordinary isotropic liquid, meaning its properties are the same in all directions. This is in contrast to the anisotropic properties of the liquid crystal in its mesotropic phases.

They have a particular molecular alignment intermediate between regular liquid and solid crystal. Several different types of order can occur in the mesotrophic phase. The most common of two are nematic and chiral nematic.

Below this temperature the liquid crystals exist in mesotrophic phase. A mesophase is a state of matter that exhibits some order and alignment of molecules, but it is not as ordered as a crystalline solid. Mesophases are characteristic of liquid crystal materials and are typically observed at temperatures below the transition temperature at which the material goes from a solid to a liquid crystal phase.
They have a particular molecular alignment intermediate between regular liquid and solid crystal.

Several different types of order can occur in the mesotrophic phase. Two of the most common are the nematic phase and the chiral nematic phase.Liquid crystalline materials possess strong nonlinear optical effects in both the isotropic and mesotropic phases .
