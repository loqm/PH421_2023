#**The Spontaneous Raman Effect**
The spontaneous Raman effect, discovered by C.V. Raman in 1928, is a phenomenon in spectroscopy where a material scatters incident light, and the scattered light has frequencies different from the excitation source. This leads to the observation of both Stokes and anti-Stokes components.


*   Stokes components: These are the Raman scattered photons that have lower frequencies (i.e., longer wavelengths) than the incident photons. They result from the absorption of energy by the sample. Stokes lines are typically more intense because thermal energy at room temperature favors transitions to lower energy states.
*   Anti-Stokes components: These are the Raman scattered photons that have higher frequencies (i.e., shorter wavelengths) than the incident photons. They result from the emission of energy by the sample. Anti-Stokes lines are typically much weaker compared to Stokes lines because they require the sample to be in a higher energy state, which is less probable at room temperature.

These properties of Raman scattering can be understood through use of the energy level diagrams shown in Fig. 10.1.2. Raman Stokes scattering consists of a transition from the ground state g to the final state n by means of a virtual intermediate level associated with excited state n′. Raman anti-Stokes scattering entails a transition from level n to level g with n′ serving as the intermediate level.
 The anti-Stokes lines are typically much weaker than the Stokes lines because, in thermal equilibrium, the population of level n is smaller than the population in level g by the Boltzmann factor exp(−ωng / kT ).

 The significance of the Raman effect in spectroscopy lies in its ability to probe transitions that would typically be forbidden in one-photon processes. For instance, in the Raman transitions depicted in Fig. 10.1.2, it is necessary for both the matrix elements $⟨g|rˆ|n′⟩$ and $⟨n′|rˆ|n⟩$ to be non-zero. This condition implies that for a material system possessing inversion symmetry and, consequently, energy eigenstates with definite parity, the initial state "g" and final state "n" must have the same parity. However, in such cases, the g → n transition is normally forbidden in single-photon electric dipole transitions because the matrix element $⟨g|rˆ|n⟩$ must unavoidably be zero.


##  **Difference between stimulated & spontaneous Raman scattering**

1.  **Efficiency :** Stimulated Raman scattering is when we use an **intense laser beam**, and **10% or more** of the energy of the incident laser beam is often converted into the Stokes frequency. In contrast, without laser beam that is in the case of spontaneous Raman scattering even for the excitation of condensed matter, the scattering cross section per unit volume for Raman Stokes scattering is only approximately **$10^{-6} cm^{-1}$**.

2. **Emission pattern :**
The spontaneous process leads to emission in the form of a **dipole radiation pattern**, whereas the stimulated process leads to emission in a **narrow cone** in the forward and backward directions.



## 💝 **Realtionship between stimulated & spontaneous Raman scattering**

👁 **Point of view :** Number of photons in various field modes.

😅 **Postulate :**  Probability of photon emitting in Stokes mode S is directly proportional to the mean number of photons per mode in the laser radiation (mL) and the mean number of photons in S by stimulated scattering (ms) and spontaneous scattering (+1).

$$
P_S=D m_L\left(m_S+1\right)
$$

D : proportionality constant (depends on the physical properties of the material medium)

We can say that rate of change of mean number of photons in S mode is probability

$$
d m_S / d t=P_S
$$

🎁 **Assumption** - Stokes mode corresponds to a wave travelling in the positive z direction at the **velocity c/n**

Therefore, the time rate change related to spatial change by,

$$
\frac{d m_S}{d z}=\frac{1}{c / n} \frac{d m_S}{d t}=\frac{1}{c / n} D m_L\left(m_S+1\right)
$$

Now we consider the two opposite limits of **mS ≪ 1** (spontaneous) and **mS ≫ 1** (stimulted raman scattering)

🎁 **Assumption** - mL is independent of z

$$
m_S(z)=m_S(0)+\frac{1}{c / n} D m_L z \quad\left(\text { for } m_S \ll 1\right)
$$

In spontaneous scattering, Stokes intensity increases in **proportion to the length of the Raman medium** and thus to the **total number of molecules contained in the interaction region**

$$
m_S(z)=m_S(0) e^{G z} \quad\left(\text { for } m_S \gg 1\right)
$$

In stimualulated scattering, **G is Raman gain coefficient** and mS(0) represents the quantum noise associated with the vacuum state

$$
G=\frac{D m_L}{c / n}
$$


✅ **Now, we can derive the relationship between the gain coefficient G for stimulated Raman scattering and the cross-section for spontaneous Raman scattering.**

The system can radiate into a large number of Stokes modes, the total rate of loss of laser photons

$$
\frac{d m_L}{d z}=-M b \frac{d m_S}{d z}=\frac{-D m_L M b}{c / n}
$$

M : total number of modes into which the system can radiate

b : accounts for the possible nonuniform angular distribution of the scattered Stokes radiation.

$$
b=\frac{\int|f(\theta, \phi)|^2 d \Omega / 4 \pi}{\left|f\left(\theta_S, \phi_S\right)\right|^2}
$$

|f(θ,φ)|2 : angular distribution of the Stokes radiation

$$
M=\frac{V \omega_S^2 \Delta \omega}{\pi^2(c / n)^3}
$$

V : volume of the region in which the modes are defined

delω : linewidth of the scattered Stokes radiation

and we know,

$$
\frac{d m_L}{d z}=-N \sigma m_L
$$

N : number density of molecules

(σ) : the cross section for Raman scattering

Therefore, D is given by

$$
D=\frac{N \sigma(c / n)}{M b}
$$

Now Gain coefficient can be given as,

$$
G=\frac{N \sigma \pi^2 c^3 m_L}{V \omega_S^2 \Delta \omega b n^3} \equiv \frac{N \pi^2 c^3 m_L}{V \omega_S^2 b n^3}\left(\frac{\partial \sigma}{\partial \omega}\right)_0
$$

We used the definition of the spectral density of the scattering cross section to express σ in terms of its line-center value (∂σ/∂ω)0 as

$$
\sigma=\left(\frac{\partial \sigma}{\partial \omega}\right)_0 \Delta \omega
$$

**The above equation gives the Raman gain coefficient in terms of the number of laser photons per mode, mL**

$$
I_L=\frac{m_L \hbar \omega_L}{A(n L / c)}=\frac{m_L \hbar \omega_L c}{V n}

Therefore, the Raman Gain Coefficient in terms of **laser intensity**,

G=\frac{N \pi^2 c^2}{\omega_S^2 b n^2 \hbar \omega_L}\left(\frac{\partial \sigma}{\partial \omega}\right)_0 I_L

# **Stimulated Raman Scattering Described by the Nonlinear Polarization**
We will now construct a conventional model, one that is non-quantum-mechanical, to explain stimulated Raman scattering. To maintain conceptual simplicity, we will limit our analysis to the scalar approximation.We assume that the optical field interacts with a vibrational mode of a molecule. We assume that the vibrational mode can be described as a simple
harmonic oscillator of resonance frequency $\omega_v$ and damping constant $\gamma$, and we denote by $\tilde{q}(t)$ the deviation of the internuclear distance from its equilibrium value $q_0$.The equation of motion
describing the molecule vibration is thus
$$\frac{d^2\tilde{q}}{dt^2} + 2\gamma\frac{d\tilde{q}}{dt} + \omega_v^2\tilde{q} = \frac{\tilde{F}(t)}{m}$$
Here, $\tilde{F}(t)$ represents any force exerted on the vibrational mode, and $m$ stands for the reduced nuclear mass.
The key assumption of the theory is that the optical polarizability of the molecule (which is typically predominantly electronic in origin) is not constant, but depends on the internuclear separation $\tilde{q}$ according to the equation:
$$\alpha(t) = \alpha_0 + \frac{\partial\alpha}{\partial q}_0\tilde{q}(t)$$

$\alpha_0$is the polarizability of a molecule in which the internuclear distance is held fixed at its equilibrium value.When the molecule is set into oscillation, its polarizability will be modulated periodically in time. Thus, the refractive index of a collection of coherently oscillating molecules will be modulated in time in accordance with the relations

$
\tilde{n}(t) = \left(1 + N \tilde{\alpha}(t)\right)^{\frac{1}{2}}.$

The temporal modulation of the refractive index will modify a beam of light as it passes through the medium. In particular, frequency sidebands separated from the laser frequency by $\pm\omega_v\$ will be impressed upon the transmitted laser beam.

Next, we examine how molecular vibrations can be driven coherently by an applied optical field. In the presence of the optical field $\tilde{E}(z, t)$, each molecule will become polarized, and the induced dipole moment of a molecule located at coordinate $z$ will be given by

$$
\tilde{p}(z, t) =\epsilon_0
 \alpha \tilde{E}(z, t).
$$

The energy required to establish this oscillating dipole moment is given by
$$
W = \frac{1}{2}\langle \tilde{p}(z, t) \cdot \tilde{E}(z, t)\rangle = \frac{1}{2} \epsilon_0 \alpha\langle \tilde{E}^2(z, t)\rangle.
$$
$$
F = \frac{dW}{dq} = \frac{\epsilon_0}{2} \frac{d\alpha}{dq}_0\langle \tilde{E}^2(z, t)\rangle
$$

The origin of stimulated Raman scattering can be understood schematically in terms of the interactions in above figures. Part (a) of the figure shows how molecular vibrations modulate the refractive index of the medium at frequency $\omega_v$ and thereby impress frequency sidebands onto the laser field. Part (b) shows how the Stokes field at frequency $\omega_S = \omega_L - \omega_v$ can beat with the laser field to produce a modulation of the total intensity of the form

$$
\tilde{I}(t) = I_0 + I_1 \cos[(\omega_L - \omega_S)t + \phi] \quad
$$

where $\phi$ is some relative phase. This modulated intensity coherently excites the molecular oscillation at frequency $\omega_L - \omega_S = \omega_v$.

The two processes depicted in sections (a) and (b) of the figure mutually enhance each other. This means that the interaction illustrated in part (b) results in a more pronounced molecular vibration. This enhanced molecular vibration, in turn, amplifies the Stokes field as shown in part (a), and subsequently, the stronger Stokes field further intensifies the molecular vibration.

To make these ideas quantitative, let us assume that the total optical field can be represented
as
$$\tilde{E}(z, t) = A_L e^{i(k_Lz - \omega_L t)} + A_S e^{i(k_Sz - \omega_S t)} + \text{c.c.}$$

the time-varying part of the applied force is then given by
$$F(z, t) = \frac{\epsilon_0}{2} \frac{\partial\alpha}{\partial q}_0 A_L A_S^* e^{i(Kz - \Delta t)} + \text{c.c.}$$
where we have introduced the notation $K = k_L - k_S$ and $\Delta = \omega_L - \omega_S.$

We adopt a trial solution of the form $$\tilde{q}(t) = q(\Delta) e^{i(Kz - \Delta t)} + \text{c.c.}$$
Plugging all this into the first equation, we get,
$$- \Delta^2 q(\Delta) - 2i\Delta\gamma q(\Delta) + \omega_v^2 q(\Delta) = \frac{\epsilon_0}{m} \frac{\partial\alpha}{\partial q}_0 A_L A_S^*.$$
and we thus find that the amplitude of the molecular vibration is given by
$$
q(\Delta) = \frac{1}{m} \frac{\partial\alpha}{\partial q}_0  \frac{A_L A_S^*}{\omega_v^2-\Delta^2 - 2i\Delta\gamma}.
$$
The Polarization is given by,
$$\tilde{P}(z, t) = N\tilde{p}(z, t) = \epsilon_0 N \left( \alpha_0 + \frac{\partial\alpha}{\partial q}_0 \tilde{q}(z, t) \right) \tilde{E}(z, t)$$

and consequently the nonlinear part of the polarization is given by
$$\tilde{P}^{NL}(z, t) = \epsilon_0 N \frac{\partial\alpha}{\partial q}_0 [q(\Delta) e^{i(Kz - \Delta t)} + \text{c.c.}]\times [A_L e^{i(k_Lz - \omega_L t)} + A_S e^{i(k_Sz - \omega_S t)} + \text{c.c.}]$$


The nonlinear polarization is observed to comprise various frequency components. The component of this expression that oscillates at the frequency ωS is referred to as the Stokes polarization and can be expressed as:
$$\tilde{P}^{NL}_S(z, t) = P(\omega_S) e^{-i\omega_S t} + \text{c.c.}$$

Its complex amplitude is given by:
$$P(\omega_S) = N\epsilon_0 \frac{\partial\alpha}{\partial q}_0 q^*(\Delta) A_L e^{ik_Sz}$$
By introducing the expression $q(\Delta)$ into this equation, we find that the complex
amplitude of the Stokes polarization is given by
$$P(\omega_S) = \epsilon_0^2 \frac{N}{m} \frac{(\partial\alpha/\partial q)_0^2 |A_L|^2 A_S}{\omega_v^2 - \Delta^2 + 2i\Delta\gamma} e^{ik_Sz}$$
We now define the Raman susceptibility through the expression
$$P(\omega_S) = 6\epsilon_0\chi_R(\omega_S) |A_L|^2 A_S e^{ik_Sz}$$.

where, for notational convenience, we have introduced $\chi_R(\omega_S)$ as a shortened form of $\chi^{(3)}(\omega_S = \omega_S + \omega_L - \omega_L)$.

we find that the Raman susceptibility is given by
$$\chi_R(\omega_S) = \epsilon_0 N\frac{\frac{(\partial\alpha/\partial q)_0^2}{6m}}{\omega_v^2 -(\omega_L - \omega_S)^2 + 2i(\omega_L - \omega_S)\gamma}$$.

The real and imaginary parts of $\chi_R(\omega_S)$ are denoted as $\chi'_R(\omega_S)$ and $\chi''_R(\omega_S)$, respectively, which are illustrated below,

10.3.3.png

Near the Raman resonance, the Raman susceptibility can be approximated as $$\chi_R(\omega_S) = \epsilon_0 N\frac{(\partial\alpha/\partial q)_0^2}{12m\omega_v}\frac{1}{[\omega_S - (\omega_L - \omega_v)] + i\gamma}$$

Note that, at the exact Raman resonance (that is, for ωS = ωL−ωv), the Raman susceptibility
is negative imaginary. (We shall see below that consequently the Stokes wave experiences
amplification.)
In order to describe explicitly the spatial evolution of the Stokes wave, we use these equations for the nonlinear polarization appearing in the driven wave equation
$$\nabla^2 \tilde{E} - \frac{\epsilon}{c^2} \frac{\partial^2 \tilde{E}}{\partial t^2} = \frac{1}{\varepsilon_0 c^2} \frac{\partial^2 \tilde{P}^{NL}}{\partial t^2}$$
We then find that the evolution of the field amplitude AS is given in theslowly varying amplitude approximation by
$$\frac{dA_S}{dz} = -\alpha_S A_S$$
where
$$\alpha_S = -3i \frac{\omega_S}{n_S c} \chi_R(\omega_S) |A_L|^2$$
is the Stokes wave “absorption” coefficient. Since the imaginary part of $\chi_R(\omega_S)$ is negative, the
real part of the absorption coefficient is negative, implying that the Stokes wave actually experiences
exponential growth. Note that αS depends only on the modulus of the complex amplitude
of the laser field. Raman Stokes amplification is thus a process for which the phase-matching
condition is automatically satisfied. Alternatively, Raman Stokes amplification can be said to
be a pure gain process or a nonparametric process.
We can also predict the spatial evolution of a wave at the anti-Stokes frequency through
use of the results of the calculation just completed.No
assumptions were made regarding the sign of $\omega_L$ − $ω_S$. We can thus deduce the form of the
anti-Stokes susceptibility by formally replacing $ω_S$ by $ω_a$ in the earlier Raman Susceptibilty equation.
$$\chi_R(\omega_a) = \frac{\epsilon_0(N/6m)(\frac{\partial\alpha}{\partial q})^2}{\omega_v^2 - (\omega_L - \omega_a)^2 + 2i(\omega_L - \omega_a)\gamma}$$
Since $ω_S$ and $ω_a$ are related through
$$ \omega_S -\omega_L=-(\omega_L-\omega_a)$$
we see that
$$\chi_R(\omega_a) = \chi_R^*(\omega_S)$$

Near Raman resonance the stokes susceptibility can be approximated as
$$\chi_R(\omega_a) = -\frac{\epsilon_0N}{12m\omega_v} \left(\frac{\partial\alpha}{\partial q}\right)^2 \left[\frac{1}{\omega_a - (\omega_L + \omega_v)+ i\gamma}\right]$$

and at the exact resonance the Raman susceptibility is positive imaginary. The amplitude of the
anti-Stokes wave hence obeys the propagation equation
$$\frac{dA_a}{dz} = -\alpha_a A_a$$
where
$$\alpha_a = \left(-3i \frac{\omega_a}{n_ac} \chi_R(\omega_a) |A_L|^2\right)$$

However, it was found experimentally (Terhune, 1963) that the anti-Stokes wave is generated
with appreciable efficiency in certain directions. The origin of anti-Stokes generation is
an additional contribution to the nonlinear polarization beyond that described by the Raman
susceptibility
Inspection of Non Linear Polarization term shows that there is a contribution to
the anti-Stokes polarization
$$P_{NL_a}(z, t) = P(\omega_a) e^{-i\omega_a t} + \text{c.c.}$$
that depends on the Stokes amplitude and which is given by
$$P(\omega_a) = N\epsilon_0 \left(\frac{\partial\alpha}{\partial q}\right)_0 q(\Delta) A_L$$=\begin{align*}
&= \frac{N\epsilon_0^2}{m} \left(\frac{\partial\alpha}{\partial q}\right)_0^2 \frac{A^2_L A^*_S}{\omega_v^2 - \delta^2 - 2i\gamma
 } e^{i(2k_L - k_S)z}
\end{align*}

(Recall that $\delta \equiv \omega_L - \omega_S = \omega_a - \omega_L$.) This contribution to the nonlinear polarization can be described in terms of a four-wave mixing susceptibility $\chi_F(\omega_a)$=$\chi_F(\omega_a) \equiv \chi^{(3)}(\omega_a = \omega_L + \Delta)$ which is defined by the relation
$P(\omega_a) = 3\epsilon_0\chi_F(\omega_a) A^2_L A^*_S e^{i(2k_L - k_S)z}$
and which is hence equal to
$$\chi_F(\omega_a) = \frac{N\epsilon_0}{3m} \left(\frac{\partial\alpha}{\partial q}\right)_0^2 \frac{1}{\omega_v^2-(\omega_L - \omega_a)^2 + 2i(\omega_L - \omega_a)\gamma}$$
we can see that
$$\chi_F(\omega_a) = 2\chi_R(\omega_a)$$

The total polarization at the anti-Stokes frequency is the sum of the contributions described by both the raman and four wave susceptibilities given by
$$P(\omega_a) = 6\epsilon_0\chi_R(\omega_a) |A_L|^2 A_a e^{ik_az} + 3\epsilon_0\chi_F(\omega_a) A^2_L A^*_S e^{i(2k_L - k_s)z}$$
The first term in this expression is automatically phase matched and and describes a nonparametric process.The second term is not automatically
phase matched and describes the parametric process of four-wave mixing.
Similarly, there is a four-wave mixing contribution to the Stokes polarization described by
$$\chi_F(\omega_S) = \frac{N\epsilon_0}{3m} \left(\frac{\partial\alpha}{\partial q}\right)_0^2 \frac{\omega_v^2}{(\omega_L - \omega_S)^2 + 2i(\omega_L - \omega_S)\gamma}$$
so that the total polarization at the Stokes frequency is given by
$$P(\omega_S) = 6\epsilon_0\chi_R(\omega_S) |A_L|^2 A_s e^{ik_sz} + 3\chi_F(\omega_S) A^2_L A^*_a e^{i(2k_L - k_a)z}$$

This four-wave mixing contribution was not included earlier because we had
not included an anti-Stokes field.The Stokes four-wave mixing susceptibility is
related to the Raman Stokes susceptibility by
$$\chi_F(\omega_S) = 2\chi_R(\omega_S)$$
and to the anti-Stokes susceptibility through
$$\chi_F(\omega_S) = \chi_F^*(\omega_a)$$
The spatial evolution of the Stokes and anti-Stokes fields is now obtained by introducing the two new polarizations to
$$\nabla^2 \tilde{E} - \frac{\epsilon}{c^2} \frac{\partial^2 \tilde{E}}{\partial t^2} = \frac{1}{\varepsilon_0 c^2} \frac{\partial^2 \tilde{P}^{NL}}{\partial t^2}$$
We assume that the medium is
optically isotropic and that the slowly varying amplitude and constant-pump approximations
are valid. We find that the field amplitudes obey the set of coupled equations
$$\frac{dA_S}{dz} = -\alpha_s A_S + \kappa_s A^*_a e^{i\Delta k z}$$
$$\frac{dA_a}{dz} = -\alpha_a A_a + \kappa_a A^*_S e^{i\Delta kz}$$
where we have introduced nonlinear absorption and coupling coefficients
$$\alpha_j = \frac{-3i\omega_j} {n_j c} \chi_R(\omega_j) |A_L|^2$$
$$\kappa_j = \frac{3i\omega_j} {2n_j c} \chi_F(\omega_j) A_L^2$$
and have defined the wavevector mismatch
$$\mathbf{k} = \mathbf{k} \cdot \hat{z} = (2k_L - k_S - k_a) \cdot \hat{z}$$

 ## **Stokes–Anti-Stokes Coupling in Stimulated Raman Scattering**



Nature of the solution to the equations describing
the propagation of the Stokes and anti-Stokes waves

$ \frac{dA_{1}}{dz} = -α_{1}A_{1} + k_{1}A^*_{2}e^{iΔkz} \tag{3.1}$

$ \frac{dA^*_{2}}{dz} = -α^*_{2}A^*_{2} + k^*_{2}A_{1}e^{-iΔkz} \tag{3.2}$


This form are commonly encountered in nonlinear optics
and also describe, for example, any forward four-wave mixing process in the constant-pump approximation.

The form of these equations suggests that we introduce the new variables
$F_{1}$ and $F_{2}$ defined by,

$ F_{1} = A_{1}e^{−iΔkz/2} \tag{3.3}$

$ F_{2}^* = A_{2}^*e^{iΔkz/2} \tag{3.4}$

We plug in these substitution in Eq. (3.1) & Eq. (3.2), we then eliminate $F_2^*$ algebraically. We then obtain a differential equation in $F_1$,

$\Big( \frac{d}{dz} + α_2^* - i\frac{Δk}{2} \Big) \Big( \frac{d}{dz} + α_1 + i\frac{Δk}{2} \Big) F_1 = k_1k_2^*F_1 \tag{3.5}$


 We assume the form of $F_1$ to be

$F_1(z) = F_1(0)e^{gz} \tag{3.6}$

where $g$ represents an unknown spatial growth rate. We substitute this form
into Eq. (3.6) and find that this equation is satisfied by the trial solution if
$g$ satisfies the algebraic equation,


$\Big( g + α_2^* - i\frac{Δk}{2} \Big) \Big( g + α_1 + i\frac{Δk}{2} \Big)  = k_1k_2^* \tag{3.7}$

In general, this equation possesses two solutions, which are given by

$g_± = -\frac{1}{2} \Big(\alpha_1 + \alpha_2^*\Big) ± \frac{1}{2}\Big[\Big(α_1 - α_2^* + iΔk\Big)^2 + 4k_1k_2^*\Big]^\frac{1}{2} \tag{3.8}$


Except for special values of $α_1, α_2, k_1, k_2, and \Delta k$, the two values of g given
by Eq. (3.8) are distinct. Whenever the two values of g are distinct we get a solution of $F_1$, thus obtaining $A_1$,

$A_1(z) = \Big(A_1^+e^{g_+z} + A_1^-e^{g_-z}\Big) e^{iΔkz/2} \tag{3.9}$

Here $A^+_
1$ and $A^−_
1$ are constants of integration whose values must be determined from the relevant boundary conditions. The general form of the solution for $A^∗_
2(z)$ is readily found,

$A^*_2(z) = \Big[\Big( \frac{g_+ + α_1 + iΔk/2}{k_1}\Big)A^+_1e^{g_+z} + \frac{g_- + α_1 + iΔk/2}{k_1}\Big)A^-_1e^{g_-z} \Big]e^{-iΔk/2} \tag{3.10}$

If we define constants $A^+_
2$ and $A^−_
2$by means of the equation,

$A_2^*(z) = \Big(A_2^{+*}e^{g_+z} + A_2^{-*}e^{g_-z}\Big) e^{-iΔkz/2} \tag{3.11}$


we see that the amplitudes $A^±_
1$ and
$A^±_
2$ are related by

$\frac{A_2^{±*}}{A_1^±} = \frac{g_± +  \alpha_1 + iΔk/2}{k_1} \tag{3.12}$

This equation shows how the amplitudes $A^+_
2$ and
$A^+_
1$ are related in the part
of the solution that grows as $exp(g_+z)$, and similarly how the amplitudes $A^-_
2$
and $A^−_
1$ are related in the part of the solution that grows as $exp(g_−z)$. We
can think of Eq. (3.12) as specifying the eigenmodes of propagation of
the Stokes and anti-Stokes waves. As written, Eq. (3.12) appears to be
asymmetric with respect to the roles of the $ω_1$ and $ω_2$ fields. However, this
asymmetry occurs in appearance only. Since $g_±$ depends on $α_1, α_2, k_1, k_2,$
and $\Delta k$, the right-hand side of Eq. (3.12) can be written in a variety of
equivalent ways, some of which display the symmetry of the interaction more
explicitly. We next rewrite Eq. (3.12) in such a manner.

By manuplilating Eq. (3.8) we can acquire,

$\Big(g_+ + \alpha_1 + \frac{iΔk}{2}\Big)\Big(g_- + \alpha_1 + \frac{iΔk}{2}\Big) = -k_1k_2^* \tag{3.13}$

$\Big(g_± + \alpha_1 + \frac{iΔk}{2}\Big) =- \Big(g_∓ + \alpha_2^* - \frac{iΔk}{2}\Big) \tag{3.14}$

Through the use of Eq. (3.13), (3.14) and (3.8), we find that the ratio $\frac{A_2^{±*}}{A_1^±}$ can be written as

$\frac{A_2^{±*}}{A_1^±} = \bigg[\frac{k_2^*\big(g_± + α_1 + iΔk/2\big)}{k_1\big(g_± + α_2^* - iΔk/2\big)}\bigg] \tag{3.15}$

this form shows explicitly the symmetry between the roles of the $ω_1$ and $ω_2$
fields.

Next, we find the form of the solution when the boundary conditions are
such that the input fields are known at the plane $z = 0$ —that is, when $A_1(0)$
and $A^∗_
2(0)$ are given. Solution for the spatial evolution
of the two interacting fields in terms of their boundary values as,


$ A_1(z) = \frac{1}{g_+ - g_-} \Big\{\Big[k_1A_2^*(0) - \Big(g_- + α_1 + \frac{iΔk}{2}\Big)A_1(0)\Big]e^{g_+z} - \Big[k_1A_2^*(0) - \Big(g_+ + α_1 + \frac{iΔk}{2}\Big)A_1(0)\Big]e^{g_-z}\Big\} e^{iΔkz/2} \tag{3.16} $


$ A_2^*(z) = \frac{1}{g_+ - g_-} \Big\{\Big[ \Big(g_+ + α_1 + \frac{iΔk}{2}\Big)A_2^*(0) + k_2^*A_1(0)\Big]e^{g_+z} - \Big[\Big(g_- + α_1 + \frac{iΔk}{2}\Big)A_2^*(0) + k_2^*A_1(0)\Big]e^{g_-z}\Big\} e^{-iΔkz/2} \tag{3.17} $

Eq. (3.17) can also be written in terms of $α_2$ instead of $α_1$.


Before applying the results of the derivation just performed to the case of
stimulated Raman scattering, let us make sure that the solution makes sense
by applying it to several specific limiting cases.

**Dispersionless, Nonlinear Medium without Gain or Loss**

For a medium without gain (or loss), we set $α_1 = α_2 = 0$. Also, for a medium
that is lossless and dispersionless,$ χ^{(3)}
(ω_1 = 2ω_0 −ω_2)$ must equal $χ^{(3)}
(ω_2 =
2ω_0 − ω_1)$, and thus the product $k_1k^∗_
2$ that appears in the solution is equal to

$ k_1k_2^* = \frac{9w_1w_2}{4n_1n_2c^2} | χ^{(3)}
(ω_2 =
2ω_0 − ω_1)|^2|A_0|^4 \tag{3.18} $

According to Eq. (3.12), the ratio of amplitudes corresponding to each eigensolution is given by

$\frac{A_2^{±*}}{A_1^±} = \frac{g_±  + iΔk/2}{k_1} \tag{3.19}$

The right-hand side of this expression simplifies considerably for the case
of perfect phase matching ($\Delta k = 0$) and becomes $±(k^∗_
2 /k_1)^{1/2}$. If we also
choose our phase conventions so that $A_0$ is purely real, we find that expression
reduces to

$\frac{A_2^{±*}}{A_1^±} =±i \Big(\frac{n_1w_2}{n_2w_1}\Big)^{1/2} ≈  ±i \tag{3.20}$

which shows that the two frequency sidebands are phased by ±π/2 radians in
each of the eigensolutions.

**Medium without a Nonlinearity**


One would expect on physical grounds that, for a medium in which $χ^{(3)}$ vanishes, the solution would reduce to the usual case of the free propagation of
the $ω_1$ and $ω_2$ waves. We now verify that our formal solution possesses this
property. By setting $k_1 = k_2 = 0$ in Eq. (3.7), and assuming for simplicity
that $\Delta k$ vanishes, we find that

$ g_+ = -α_2^* ; g_- = -α_1^* \tag{3.21}$

The eigenamplitudes are found most readily from Eq. (3.15). If we assume
that $k_1$ and $k_2$ approach zero in such a manner that $k^∗_
2 /k_1$ remains finite, we
find from Eq. (3.15) that

$\frac{A_2^{+*}}{A_1^+} = ∞  ;  \frac{A_2^{-*}}{A_1^-} = 0 \tag{3.22}$

Thus, the positive root corresponds to a wave at frequency $ω_2$, which propagates according to

$A_2^*(z) = A_2^*(0)e^{g_-z} = A_2^*(0)e^{-α_2^*}(z) \tag{3.23}$

whereas the negative root corresponds to a wave at frequency $ω_1$, which propagates according to


$A_1(z) = A_1(0)e^{g_-z} = A_1(0)e^{-α_1}(z) \tag{3.24}$

We thus see that each of the waves simply experiences free propagation.

```python
#Dispersionless, Nonlinear Medium without Gain or Loss
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the coupled differential equations
def coupled_equations(y, z, chi_S_real, chi_S_imag, chi_A_real, chi_A_imag):
    AS_real, AS_imag, AA_real, AA_imag = y
    dAS_real_dz = chi_S_imag * AL**2 * AA_real
    dAS_imag_dz = -chi_S_real * AL**2 * AA_imag
    dAA_real_dz = chi_A_imag * AL**2 * AS_real
    dAA_imag_dz = -chi_A_real * AL**2 * AS_imag
    return [dAS_real_dz, dAS_imag_dz, dAA_real_dz, dAA_imag_dz]

# Initial conditions
AS0_real = 1.0
AS0_imag = 0.0
AA0_real = 0.0
AA0_imag = 0.0
y0 = [AS0_real, AS0_imag, AA0_real, AA0_imag]

# Parameters
AL = 1.0  # Pump wave amplitude
chi_S_real = 0.0  # Real part of the nonlinear coefficient for the signal
chi_S_imag = 1.0  # Imaginary part of the nonlinear coefficient for the signal
chi_A_real = 0.0  # Real part of the nonlinear coefficient for the idler
chi_A_imag = 0.8  # Imaginary part of the nonlinear coefficient for the idler


z = np.linspace(0, 10, 1000)

solution = odeint(coupled_equations, y0, z, args=(chi_S_real, chi_S_imag, chi_A_real, chi_A_imag))

# Extract the solutions for AS and AA
AS_real, AS_imag, AA_real, AA_imag = solution.T

# Create plots
plt.figure(figsize=(10, 6))
plt.plot(z, np.abs(AS_real + 1j * AS_imag), label='|AS|')
plt.plot(z, np.abs(AA_real + 1j * AA_imag), label='|AA|')
plt.xlabel('z')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Dispersionless Nonlinear Medium without Gain or Loss')
plt.grid()
plt.show()
```

# **Stokes-Anti-Stokes Coupling in Stimulated Raman Scattering**

We will use the solutions to the coupled differential equations solved in the above sections and apply it in analysing the case of stimulated Raman scattering.

The nonlinear absorption coefficients are represnted by $α_s$ and $α_a$ and coupling coefficients by $κ_s$ and $κ_a$ whereas the frequencies associated with the stokes and anti-stokes components are represented by $ω_s$ and $ω_a$.

We now use the relation from previous sections that connect the four wave mixing susceptibilities with the raman susceptibilities.

\begin{equation}
χ_F\left(ω_s\right) = χ_F\left(ω_a\right)^* = 2χ_R\left(ω_s\right) = 2χ_R\left(ω_a\right)^*  
\end{equation}




We will also be using the relations among coupling coeffecients and absorption coefficients as described below
\begin{equation}
α_a = -α_s\left(n_s ω_a/n_a ω_s  \right)
\end{equation}
\begin{equation}
\kappa_s = -\alpha_s e^{2\iota ϕ_L}
\end{equation}
\begin{equation}
\kappa_a = α_s^*\left(n_s ω_a/n_a ω_s \right) e^{2\iota ϕ_L}
\end{equation}

where $ϕ_L$ is the phase of the pump laser desrcibed in the equation $A_L = |A_L|e^{\iota ϕ_L}$. The stokes absorption coefficient is given by
\begin{equation}
\alpha_s = \frac{-ιω_s N ϵ_{0}^2 \left(\frac{∂\alpha}{∂q}\right)_{0}^2|A_L|^2 }{4mnc\omega_v[\omega_s - \left(\omega_L - ω_v\right) + \iota γ]}
\end{equation}

This sets up the environment to write the eigenvalues for the coupled differential equations in terms of $\alpha_s$ and following the results from previous section, the final form is deduced as follows

\begin{equation}
g_+ = -1/2\alpha_s\left(1 - n_sω_a/n_a\omega_s\right) + 1/2[[α_s\left(1 + n_sω_a/n_a\omega_s\right) + \iota Δk]^2 - 4α_s^2 n_sω_a/n_a\omega_s]^{1/2}
\end{equation}
\begin{equation}
g_- = -1/2\alpha_s\left(1 - n_sω_a/n_a\omega_s\right) - 1/2[[α_s\left(1 + n_sω_a/n_a\omega_s\right) + \iota Δk]^2 - 4α_s^2 n_sω_a/n_a\omega_s]^{1/2}
\end{equation}

Under the approximation $n_sω_a/n_a\omega_s = 1$, the eigenvalues simplify to
\begin{equation}
g_+ = [\iota α_s Δk - \left(Δk/2\right)^2]^{1/2}
\end{equation}
\begin{equation}
g_- = -[\iota α_s Δk - \left(Δk/2\right)^2]^{1/2}
\end{equation}

we can see from the graph that for very small phase mismatch, both $g_+$ and $g_-$ vanishes because the antistokes and stokes wave are strongly coupled to each other so much such that it also prevents stokes wave from experiencing any growth.

**In the limit $Δk$ very large**

The eigenvalues approximately become
\begin{equation}
g_+ = \left(\alpha_s + 1/2 \iota \Delta k\right)
\end{equation}
\begin{equation}
g_- = -\left(\alpha_s + 1/2 \iota \Delta k\right)
\end{equation}

and the sidemode amplitude ratios associated with each of these eigenvalues could be approximated as
\begin{equation}
\frac{A_{a}^{+*}}{A_{s}^+} = ιΔk/α_s
\end{equation}
\begin{equation}
\frac{A_{a}^{-*}}{A_{s}^-} = 0
\end{equation}


**In the limit $Δk$ very small**

The eigenvalues under the condition $\Delta k = 0$ approximately become
\begin{equation}
g_+ = 0
\end{equation}
\begin{equation}
g_- = -α_s \left(1 - n_s \omega_a/n_a \omega_s\right)
\end{equation}

and the sidemode amplitude ratios associated with each of these eigenvalues could be approximated as
\begin{equation}
\frac{A_{a}^{+*}}{A_{s}^+} = -1
\end{equation}
\begin{equation}
\frac{A_{a}^{-*}}{A_{s}^-} = -1
\end{equation}

Hence we can see that each eigensolution is an equal combination of stokes and anti-stokes components.


Now in this limit, we will look at the spatial evolution of field amplitudes with proper boundary conditions at z = 0.

\begin{equation}
A_s\left(z\right) = \frac{-1}{1 - n_s \omega_a /n_a ω_s} \{[A_{a}^*\left(0\right)e^{2\iota ϕ_L} + n_s \omega_a /n_a ω_s A_s\left(0\right)] - [A_{a}^*\left(0\right)e^{2\iota ϕ_L} + A_s\left(0\right)] e^{g_{-}z}\}
\end{equation}

\begin{equation}
A_a\left(z\right)^* = \frac{1}{1 - n_s \omega_a /n_a ω_s} \{[A_{a}^*\left(0\right) + n_s \omega_a /n_a ω_s A_s\left(0\right) e^{-2\iota ϕ_L}] - n_s \omega_a /n_a ω_s [A_{a}^*\left(0\right) + A_s\left(0\right)e^{-2\iota ϕ_L}] e^{g_{-}z}\}
\end{equation}

The negative nature of $g_-$ implies that the second term of each expression exponentially decays with z and the asymptotic values of the field amplitudes are then given as follows

\begin{equation}
A_s\left(z → ∞\right) = \frac{-1}{1 - n_s \omega_a /n_a ω_s} [A_{a}^*\left(0\right)e^{2\iota ϕ_L} + n_s \omega_a /n_a ω_s A_s\left(0\right)]
\end{equation}


\begin{equation}
A_{a}^*\left(z → ∞\right) = \frac{1}{1 - n_s \omega_a /n_a ω_s} [A_{a}^*\left(0\right) + n_s \omega_a /n_a ω_s A_s\left(0\right)e^{-2\iota ϕ_L}]
\end{equation}

Each field is amplified by the factor ${\left(1 - n_s \omega_a /n_a ω_s\right)^{-1}}$ but after several times $1/g_-$ distance the amplitudes saturate to a constant value and then the amplitudes no longer change with the distance.

The reason to why amplitudes attain steady value is governed by the whole exercise we did earlier by solving the vibrational model of the molecule but now in the simultaneous presence of 3 fields (stokes, anti-stokes and laser)

\begin{equation}
E\left(z,t\right) = A_L e^{k_L z - ω_L t} + A_s e^{k_s z - ω_s t} + A_a e^{k_a z - ω_a t} + c.c.
\end{equation}

Now the solution to the model equation of a vibrating molecule with this modified source term gives solution (inter-nuclear distance q) of the form

\begin{equation}
q\left(z,t\right) = q\left(Ω\right) e^{\iota \left(Kz - \Omega t\right)} + c.c.
\end{equation}

where $K = k_L -k_s = k_a - k_L$, $\omega_L - \omega_s = \omega_a - \omega_L = \Omega$ and

\begin{equation}
q\left(Ω\right) = \frac{\left(\epsilon_0 /m\right) \left(\partial \alpha / \partial q\right)_0 \left(A_L A_{s}^* + A_a A_{L}^*\right)}{\omega_{v}^2 - \Omega^2 - 2 \iota \Omega γ}
\end{equation}

#**Coherent Anti-Stokes Raman Scattering**

In this section we will analyse Coherent Anti-Stokes Raman scattering (CARS) and Coherent Stokes Raman scattering (CSRS). We will mostly focus on CARS, where usually 2 beams of different frequecies $𝜔_1$ and $𝜔_2$ are passed through a Raman medium and due to this interaction a new beam at a different frequency is created. In the process of Coherent anti-Stokes Raman scattering, an output is created at frequency 2$𝜔_1$ - $𝜔_2$ as a consequence of the susceptibility *$𝜒^{(3)}_F (ω_a = ω_1 - ω_2 + ω_1) $*.

Let us further more carefully analyse coherent anti-Stokes Raman scattering (CARS). From the prevuious sections we now know that the susceptibility describing the process is

$𝜒^{(3)}_F (ω_a = ω_1 − ω_2 + ω_1) = \frac{−(ϵN/6mω_v)(∂α/∂q)^{2}_0}{
[(ω_1 − ω_2) − ω_v] + iγ } \tag{5.1}$

the nonlinear response experiences a resonance whenever the input frequencies ω1
and ω2 are selected so that $𝜔_1$ - $𝜔_2$ is equal to a vibrational frequency $𝜔_v$ of the material system.

We can write the coupled-amplitude equation for the generation of the anti-Stokes wave as

$\frac{{dA_a}}{dz} = −α_{a}A_{a} + 𝜅_{a}A_{2}^{*}e^{iΔkz} \tag{5.2}$

where

$α_{a}=\frac{-3i𝜔_a}{η_{a}c}χR(𝜔_a)|A_1|^2, 𝜅_{a}=\frac{3i𝜔_a}{2η_{a}c}χR(𝜔_a)A_1^2 \tag{5.3}$


image.png

_The above graph is that of a typical CARS lineshape_

and

$Δk = Δk.\hat{z} = (2k_{1}-k_{2}-k_{a}).\hat{z} \tag{5.4}$

CARS is usually studied under conditions such that the generation of the anti-Stokes signal
is dominated by the second term on the right-hand side of Eq. (5.2). This situation occurs
whenever the gain of stimulated Raman scattering (described by the first term) is small. Under
this assumption and that of perfect phase matching (Δk = 0), the growth of the anti-Stokes
wave can be expressed as

$A_{a}=\frac{3i𝜔_a}{2η_{a}c}χF(𝜔_a)A_1^2A_2^{*}z \tag{5.5}$

The spectral variation of the anti-Stokes generation as either of the two input frequencies is
varied is determined by $|χF(𝜔_a)|^2$, and thus is expected to have a shape in the form of the square of a Lorentzian function. However, it is found experimentally that the measured line shape is often highly distorted. An example of such a line shape is shown in Fig. (5.2). The explanation for this behavior is that in addition to the highly resonant molecular response described by
Eq. (5.1) actual materials also exhibit a nonresonant background susceptibility $χ^{(3)}_{NR}$ due to electronic response and to nonresonant vibrational modes. The total susceptibility is then given
by $χ^{(3)}_{NR} +χ^{(3)}_F(𝜔_a)$, and consequently the lineshape will be given by $|χ^{(3)}_{NR} +χ^{(3)}_F(𝜔_a)|^2$, which leads to a line shape of the sort shown in the figure.
