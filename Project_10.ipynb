# Theoretical Derivation of SPDC: Classical Case

The idea is that when an electromagnetic wave interacts with a medium full of charges, there is
a dipolar-type of interaction between the dipoles (charges) in the medium and the incoming EM field
at the frequency $\omega$. The classical harmonic oscillator model holds which is the model of an electron
attached to an atom-core behaving like a spring and thus like an harmonic oscillator interacting with
the incoming wave. We then obtain the standard linear complex refractive index of the medium, which
would depend on œâ. But if one starts to increase significantly the energy of the incoming field then this
harmonic oscillator will start to be anharmonic i.e. not respond linearly with the incident
electromagnetic wave. We evaluate such a system by considering a non centrosymmetric potential $U = \frac{1}{2}kx^2 +\frac{1}{3}max^3$. The presence of this potential leads to non-linear effects since the resultant equation of motion is a non-linear differential equation.


$$ \frac{d^2 x}{d t^2} + 2 \gamma \frac{dx}{d t} + \omega_0^2 x + a x^2 = F = - \frac{e}{m} E(t)$$

Note that we have approximated the electric field to be constant in space due to the miniscule scale of length that we are operating at. This differential equation is solved using perturbation theory, and we obtain expressions for the first few orders of displacement.

## Solving the Optical Wave Equation

On manipulating Maxwell's equations we arrive the Optical Wave Equation. This is of interest to us since we can obtain the amplitudes of input and output waves by working on these equations when examining phenomena such as wave propagation inside a non-linear crystal. The equation is as follows:

$$ \nabla^2 \overrightarrow{E} - \frac{1}{c^2} \frac {\partial^2 \overrightarrow{E}}{\partial t^2} = \frac{1}{\epsilon_0 c^2} \frac {\partial^2 \overrightarrow{P}}{\partial t^2} $$

Note that $\overrightarrow{P} = \overrightarrow{P}^L + \overrightarrow{P}^{NL}$. SPDC is a second order non linear phenomena which involves three waves, and therefore, we will only examine effects up to the second order.
The first non-linear polarisation term is given as:

$$
P_i^{NL} = \epsilon_0 \Sigma_j \Sigma_k \chi_{ijk}^{(2)} E_j E_k
$$

with {ùëñ, ùëó, ùëò} = {ùë•, ùë¶, ùëß}, meaning that the second order term $\chi_{ijk}^{(2)}$ is actually a tensor and is called the
second order nonlinear susceptibility of the medium.

The SPDC process 'down-converts' an $\omega_3$ photon into two resultant $\omega_1$ and $\omega_2$ photons while obeying the laws of conservation of energy and momentum.

$$ \hbar \omega_3 = \hbar \omega_2 + \hbar \omega_1 $$
$$ \overrightarrow{k_3} = \overrightarrow{k_2} +  \overrightarrow{k_1}$$
$$ \implies \Delta k = \overrightarrow{k_3} - \overrightarrow{k_2} -  \overrightarrow{k_1} = 0$$

These are the **phase-matching** conditions.

We proceed to solve it by considering an ansatz of the following form:

$$ \overrightarrow{E}_i(z, t) = \overrightarrow{A}_ie^{i(k_iz - w_it)} + c.c $$

Note that in the above expression $k_3$ = $n_3 \omega_3/c$ and $n_3^2 = \chi_3^{(1)}$. As a consequence of the above expression, the polarisation of the $\omega_3$ wave is of the form $P_3(z, t) = 4 \epsilon_0 d_{eff} E_1 E_2e^{(k_3z - w_3t)} + c.c$

Using the ansatz mentioned, we essentially convert the wave equation from a time-domain equation a frequency-domain equation by considering $E$ as a superposition of constituent waves of different frequency and then taking the time derivatives. Note that this holds for the polarisation as well.
The frequency domain optical wave equation thus obtained is:

$$ \nabla^2 \overrightarrow{E} - \frac{\omega^2}{c^2}\epsilon \overrightarrow{E} = - \mu_0 \omega^2 \overrightarrow{P}^{NL}(\omega, z) $$

Into this, we substitute our electric field ansatz while keeping in mind the **slowly varying envelope approximation**. This allows us to neglect the second spatial derivative of the electric field amplitude with respect to the wavelength of the wave. On plugging in the ansatz, we obtain the following:

$$
2ik_3 \frac{dA_3}{dz} = - \frac{4d_{eff}\omega_3}{c^2} A_1 A_2 e^{i\Delta kz}
$$

where $\Delta k = k_1 + k_2 - k_3$. We can integrate this expression with respect to $z$ considering the limits to be $0$ to $L$ i.e the beam has travelled the length of the crystal. We then relate $A_3$ to the intensity of the beam $I_3$ using $I_i = 2n_i\epsilon_0 c |A_i|^2$. This yields:

$$
I_3 = \frac{8 d^2_{eff} \omega_3^2 I_1 I_2}{n_1 n_2 n_3 \epsilon_0 c^3} L^2 (\frac{sin \Delta k L/2}{ \Delta k L/2})^2 = I_3^{max} sinc^2(\Delta kL/2)
$$

To understand the variation of $I_3$ with $\Delta k$, we plot a schematic graph of the $sinc^2$ function:

```python
# Generate x values
x = np.linspace(-10, 10, 400)  # Adjust the range and number of points as needed

# Calculate y values
y = np.sinc(x)
y = y*y

# Plot the graph
plt.plot(x, y, label='sin^2(x)/x', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin^2(x)/x')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

This expression is maximised when $\Delta k = 0$ i.e when perfect phase matching is achieved. This means that:

$$
k_1 + k_2 = k_3
$$

$$
\implies n_1\omega_1/c + n_2 \omega_2/c = n_3 \omega_3/c
$$

Satisfying this equation is not possible under conditions of an ordinary crystal as most materials have $n_1(\omega_1) < n_2(\omega_2) < n_3(\omega_3)$ for $\omega_1 < \omega_2 < \omega_3 $. This is where birefringent crystals are used as they
possess two (uniaxial) or three (biaxial) different refractive indices along different symmetry axes
which can satisfy the phase-matching condition. In the particular case of second-harmonic generation,
it is fairly easy to have $n_e(\omega_1) = n_o(\omega_3 = 2\omega_1)$ with o and e the so-called ordinary and extraordinary
axes for uniaxial non-linear crystals.

SPDC is the reverse process of sum-frequency generation (SFG which is non-degenerate SPDC, i.e.
different wavelengths for the two output photons) or SHG (degenerate SPDC, i.e. same wavelengths
for the two output photons) which would correspond to the process called parametric amplification in
classical non-linear optics. This process occurs when a strong pump beam at $\omega_3$ is combined with a
weak signal beam at $\omega_1$ to give rise to an amplified signal at $\omega_1$ but also to a second signal called the
idler beam at $\omega_2$. Using the same type of equation than the one derived earlier, we have this time:

$$
\frac{dA_1}{dz} = i \frac{\omega_1}{n_1 c} \chi_{eff}^{(2)} A_3 A_2^* e^{i\Delta k z}
$$

$$
\frac{dA_2}{dz} = i \frac{\omega_2}{n_2 c} \chi_{eff}^{(2)} A_3 A_1^* e^{i\Delta k z}
$$

To solve these coupled differential equations, we solve them by considering them in matrix form, then diagonalising the operator after converting it to a stationary basis and writing the general solution as a sum of the two eigenfunctions. This gives:

$$
A_2(z) = i \sqrt{\frac{\omega_2 n_1}{\omega_1 n_2}} \frac{A_3}{|A_3|} A_1^*(0) sinh(\alpha z)
$$

$$
A_1(z) = A_1(0) cosh(\alpha z)
$$

$$
\alpha = \frac{\chi_{eff}^{(2)} |A_3|}{c} \sqrt{\frac{\omega_2 \omega_1}{n_1 n_2}}
$$

From these equations, we clearly notice why SPDC is not permitted by the classical theory of optics. The spontaneity of SPDC implies that the phenomenon must happen on it's own without the presence of any other factors. However these equations show that if there is no initial incident power at frequency $\omega_1$ i.e if $A_1(0) = 0$, then the signal and idler beams would not exist and SPDC would not be possible.

#Derivation of Quantised Electromagnetic Field

To quantise the electromagnetic field, we start with Maxwell's equations:

\begin{equation}
    \nabla \times \textbf{E} + \frac{\partial \textbf{B}}{\partial t} = 0
\end{equation}

and

\begin{equation}
    \nabla \cdot \textbf{B} = 0
\end{equation}

We use Coulomb gauge i.e. $\nabla \cdot A = 0$ which results in:

\begin{equation}
    \textbf{B} = \nabla \times \textbf{A}
\end{equation}

and

\begin{equation}
    \textbf{E} = -\frac{\partial \textbf{A}}{\partial t}
\end{equation}

Making these substitutions, we get

\begin{equation}
  \bigg( \nabla^{2} - \frac{1}{c^{2}} \frac{\partial^{2}}{\partial t^{2}} \bigg) \textbf{A} = 0
\end{equation}

The solution of which is given by:

\begin{equation}
  \textbf{A}(\textbf{r},t) = (\textit{A} e^{i(\textbf{k} \cdot \textbf{r} - \omega t)} + c.c. ) \textbf{e}
\end{equation}

where $\textit{A}$ is a complex number denoting the complex amplitude and phase of wave, c.c represents the complex conjugate of the first term and $\textbf{e}$ denotes a unit vector in the direction of $\textbf{A}$.<br />
Also,

\begin{equation}
  \omega = c|\textbf{k}|
\end{equation}

and

\begin{equation}
  \nabla \cdot \textbf{A} = 0 \implies \textbf{k} \cdot \textbf{e} = 0
\end{equation}

Suppose we put electromagnetic field in box of volume $V = L^{3}$, having periodic boundary conditions (later on, we will take the limit $L \to \infty$). Thus, the components of $\textbf{k}$ form a discrete set:

\begin{equation}
  k_{j} = \frac{2\pi n_{j} } {L}, n_{j} \in \textbf{Z}, j = 1,2,3
\end{equation}

We then write the vector polarisation as:

\begin{equation}
  \textbf{A}(\textbf{r},t) = \sum_{\textbf{k}\lambda} \big ( A_{\textbf{k} \lambda} e^{(i(\textbf{k} \cdot \textbf{r} - \omega t)} + c.c \big ) \textbf{e}_{\textbf{k}\lambda}
\end{equation}

where $\lambda$ denotes the two possible directions of polarisation for each wave vector $\textbf{k}$ and hence, $\lambda = 1,2$.
The classical Hamiltonian for electromagnetic field is given by:

\begin{align}
  \hat{H} &= \frac{1}{2} \int \big ( \epsilon_{o} |\textbf{E}|^{2} + \frac{1}{\mu_{o}} |B|^{2} \big) d^{3}\textbf{r} \\
  &= \frac{1}{2} \int \bigg( \epsilon_{o} \bigg| \frac{\partial{A}}{\partial t} \bigg|^{2} + \frac{1}{\mu_{o}} |\nabla \times \textbf{A}|^{2}\bigg) d^{3}\textbf{r}
\end{align}

This can be simplified by using Parseval's theorem which states that:

\begin{equation}
  \int |f(\textbf{r})|^{2} d\textbf{r} = \sum_{k} |\tilde{f}(\textbf{k})|^{2}
\end{equation}

(upto constant factors). Substituting these expressions and simplifying, we finally get,

\begin{equation}
  \hat{H} = 2\epsilon_{o} \sum_{\textbf{k},\lambda} \omega_{k}^{2} (|\textit{A}_{\textbf{k}, \lambda}^{R}|^{2} + |\textit{A}_{\textbf{k}, \lambda}^{I}|^{2})
\end{equation}

We write $\textit{A}_{\textbf{k},\lambda}(t)$ as $\textit{A}_{\textbf{k},\lambda}(t) = A_{\textbf{k}\lambda} e^{-i\omega_{\textbf{k}}t}$. Note that,

\begin{align}
  \dot{A}_{\textbf{k},\lambda}^{R} &= \omega_{\textbf{k}}A_{\textbf{k},\lambda}^{I} \\
  \dot{A}_{\textbf{k},\lambda}^{I} &= -\omega_{\textbf{k}}A_{\textbf{k},\lambda}^{R}
\end{align}

So,

\begin{align}
  \frac{\partial H}{\partial A_{\textbf{k},\lambda}^{R}} &= 4\epsilon_{o}\omega_{\textbf{k}}^{2}A_{\textbf{k},\lambda}^{R} = -4\epsilon_{o}\omega_{\textbf{k}} \dot{A}_{\textbf{k},\lambda}^{I} \\
  \frac{\partial H}{\partial A_{\textbf{k},\lambda}^{I}} &= 4\epsilon_{o}\omega_{\textbf{k}}^{2}A_{\textbf{k},\lambda}^{I} = 4\epsilon_{o}\omega_{\textbf{k}} \dot{A}_{\textbf{k},\lambda}^{R}
\end{align}

Defining

\begin{align}
  Q_{\textbf{k},\lambda} &= 2 \sqrt{\epsilon_{o}} A_{\textbf{k},\lambda}^{R} \\
  P_{\textbf{k},\lambda} &= 2 \omega_{\textbf{k}} \sqrt{\epsilon_{o}} A_{\textbf{k},\lambda}^{I}
\end{align}

These satisfy the relation:

\begin{align}
  \frac{\partial H}{\partial Q_{\textbf{k},\lambda}} &= -\dot{P}_{\textbf{k},\lambda} \\
  \frac{\partial H}{\partial P_{\textbf{k},\lambda}} &= \dot{Q}_{\textbf{k},\lambda}
\end{align}

The Hamiltonian then becomes similar to that of harmonic oscillator:

\begin{equation}
  H = \frac{1}{2} \sum_{\textbf{k},\lambda} (P_{\textbf{k},\lambda}^{2} + \omega_{\textbf{k}}^{2} Q_{\textbf{k},\lambda}^{2})
\end{equation}

We then promote $P_{\textbf{k},\lambda}$ and $Q_{\textbf{k},\lambda}$ to operators $\hat{P}_{\textbf{k},\lambda}$ and $\hat{Q}_{\textbf{k},\lambda}$ and define ladder operators:

\begin{align}
  a_{\textbf{k},\lambda}^{\dagger} &= \sqrt{ \frac{\hslash}{2\omega_{\textbf{k}}} } (\omega_{\textbf{k}} \hat{Q}_{\textbf{k}, \lambda} - i\hat{P}_{\textbf{k}, \lambda}) \\
  a_{\textbf{k},\lambda} &= \sqrt{ \frac{\hslash}{2\omega_{\textbf{k}}} } (\omega_{\textbf{k}} \hat{Q}_{\textbf{k}, \lambda} + i\hat{P}_{\textbf{k}, \lambda})
\end{align}

which on substituting finally gives:

\begin{equation}
  \hat{H} = \sum_{\textbf{k}, \lambda} \hslash \omega_{\textbf{k}}  \bigg ( a_{\textbf{k}, \lambda}^{\dagger}a_{\textbf{k}, \lambda} + \frac{1}{2} \bigg)
\end{equation}

This is the quantised free radiation Hamiltonian.



#Theoretical Derivation of SPDC: Quantum Case

As we have seen earlier, classical derivations fall short in explaining SPDC, and we need to employ Quantum Optics to understand why SPDC occurs.

In SPDC, idler and signal beams spontaneously appear without any input; a photon pair suddenly materializes seemingly out of nowhere. To explain this phenomenon, we need to quantize the electromagnetic field. The quantization of electromagnetic radiation leads to the discretization of energy levels into photon states. We work with the vector potential $\mathbf{A}$ instead of the electric field $\mathbf{E}$, specifically in the Coulomb gauge, where $\nabla \cdot \mathbf{A} = 0$.

$$
\vec{A}(\mathbf{r}) = \sum_l \vec{\epsilon}_l\frac{\mathcal{E}_l}{\omega_l} \left(e^{i\mathbf{k}_l \cdot \mathbf{r}} \hat{a}_l + e^{-i\mathbf{k}_l \cdot \mathbf{r}} \hat{a}_l^\dagger\right) = \vec{A}^{(+)}(\mathbf{r}) + \vec{A}^{(-)}(\mathbf{r})
$$

Here, $\hat{a}_l^\dagger$ and $\hat{a}_l$ are the creation and annihilation operators of a quantum of energy, which in the context of quantum optics are referred to as photons. The following relations apply:

$$
\hat{a}_l^\dagger |0_l\rangle = |1_l\rangle, \quad \hat{a}_l |1_l\rangle = |0_l\rangle
$$

The quantized free radiation Hamiltonian that we derived earlier is given by:

$$
\hat{H}_R = \sum_l \hbar \omega_l \left(\hat{N}_l + \frac{1}{2}\right) = \sum_l \hbar \omega_l \left(\hat{a}_l^\dagger \hat{a}_l + \frac{1}{2}\right)
$$

Here, $\hat{N}_l$ is the photon number operator, representing the number of photons in electromagnetic mode $l$.

For the SPDC phenomenon, where a high-energy photon spontaneously splits into two lower-energy photons when interacting with a nonlinear medium, the effective interaction Hamiltonian looks like:

$$
\hat{H}_{SPDC} = i\hbar\kappa \left(\hat{a}_i \hat{a}_s \hat{a}_p^\dagger e^{i\Delta\mathbf{k} \cdot \mathbf{r} - i\Delta\omega t} + \hat{a}_i^\dagger \hat{a}_s^\dagger \hat{a}_p e^{-i\Delta\mathbf{k} \cdot \mathbf{r} + i\Delta\omega t}\right)
$$

where $\kappa = \frac{2}{3} \frac{d_{eff}}{\epsilon_0 V} \sqrt{\frac{\omega_p \omega_s \omega_i}{2 \epsilon_0 V}}$

The first term represents SFG or SHG (if $\omega_i = \omega_s$) with $\Delta\omega = \omega_p - \omega_i - \omega_s$, accounting for the annihilation of idler and signal beams and the creation of the pump beam. The second term represents SPDC, where the pump beam is annihilated, giving rise to idler and signal beams. We will focus on degenerate SPDC, where $\omega_i = \omega_s$

Let's describe the situation in Quantum Mechanics. We denote our wavefunction as $|N_s, N_i, N_p\rangle$, where $N$ represents the number of photons in the corresponding beams. Initially, our state is $|0_s, 0_i, N_p\rangle$, as only pump photons are initially present. Then the SPDC Hamiltonian interacts with the state, and it evolves according to the Schr√∂dinger equation. This yields the state at time $t$ as:

$$
|\Psi(t)\rangle = e^{\frac{1}{i\hbar} \int \hat{H}_{SPDC}(t') dt'} |0_s, 0_i, N_p\rangle
$$

As $\Delta w = 0$ in this case, our Hamiltonian becomes time independant, thus:

$$
|\Psi(t)\rangle = e^{\kappa t \hat{a}_i \hat{a}_s \hat{a}_p^\dagger e^{i\Delta\mathbf{k} \cdot \mathbf{r}} + cc.} |0_s, 0_i, N_p\rangle
$$

Let $\kappa t << 1$. Using a Taylor expansion, we get:

$$
|\Psi(t)\rangle = C |0_s, 0_i, N_p\rangle + \kappa t C(\hat{a}_i \hat{a}_s \hat{a}_p^\dagger e^{i\Delta\mathbf{k} \cdot \mathbf{r}} + \hat{a}_i^\dagger \hat{a}_s^\dagger \hat{a}_p e^{-i\Delta\mathbf{k} \cdot \mathbf{r}})|0_s, 0_i, N_p\rangle
$$
<br>
$$
\implies |\Psi(t)\rangle = C|0_s, 0_i, N_p\rangle + \kappa t Ce^{i\Delta\mathbf{k} \cdot \mathbf{r}}|1_s, 1_i, N_p -1\rangle
$$

where $ùê∂$ is the normalization factor of the wavefunction.

The measured electromagnetic (EM) intensity, denoted as $I_{\text{SPDC}}$, is directly proportional to the square of the probability amplitude represented by the wave function. This wave function characterizes the likelihood of being in the SPDC state (wherein signal and idler photons are generated). Consequently, the phase matching condition becomes evident in its expression. Unlike Second Harmonic Generation (SHG), SPDC exhibits a linear relationship with respect to the power of the pump beam. Assuming a sufficiently strong pump power, it can be approximated as a classical EM field, $a_{\text{p}} \approx E_{\text{p}}$, resulting in $I_{\text{SPDC}}$ being directly proportional to $I_{\text{p}}$ alone, unlike the SHG process. In SPDC, a single pump photon gives rise to two photons, one in each mode, making the dependence clearly linear, despite the underlying non-linear nature of the process. It's worth noting that the coefficient of the SPDC state is significantly smaller, primarily due to the factor $\kappa t << 1$. This inefficiency in SPDC is one of its major drawbacks.

Furthermore, we've only considered the first two terms of the Taylor expansion. If the approximation $a_{\text{p}} \approx E_{\text{p}}$ holds and even stronger pump beam is given, higher-order terms come into play. These higher-order terms give rise to states in which multiple photon pairs are created. Consequently, if the aim is to avoid generating multiple pairs, excessive pump intensity should be avoided.

In practical quantum optics experiments, a technique known as coincidence measurement is employed to detect SPDC photon pairs. This technique involves simultaneous detections on two detectors within a time window of approximately 10 ns. This approach effectively eliminates the contribution of vacuum state detections (which are a significant part of the overall wavefunction) and focuses on the state associated with one or more photon pairs.

An important observation, as demonstrated by Mandel et al., is that the number of SPDC pair events created must not exceed the number of incoming photons, assuming that each incoming photon is converted without higher-order effects.

In conclusion, SPDC's existence can be established only through quantum formalism, as classical physics alone cannot account for this phenomenon.

## Phase Matching of SPDC

**SPDC** is a process where a single _pump_ photon splits into two daughter photons conventionally known as the _signal_ and _idler_ photons.  
As this process is seeded only by the electromagnetic vacuum, it is a purely quantum process, and in fact the photons created by this process possess strong quantum properties.  

The properties of SPDC differ for type-I and type-II phase matching.  
1. **Type-I**  
    - We consider this process happening in a negative uniaxial crystal for degenerate downconversion ($w_i = w_s$)
    - We rotate the crystal gradually changing the phase-matching condition.
    - At small pump angle $\theta$, the collinear wavevector mismatch $\Delta k = k_1+k_2-k_3 = n_1w_1+n_2w_2-n_3w_3 < 0$. Hence no downconverted light is observed
    -  When we increase the angle to the exact phase matching angle, for which $\Delta k=0$ we observe downconverted light only at the axis of the pump beam (collinear phase matching)
    - Now if we increase the angle further we find phase matching is acheived under noncollinear conditions
    - Hence we find that emitted light is in the form of a cone surrounding the transmitted pump beam
    - This light has string quantum properties

```python
def ellipse(a,b,style="solid", col="black"):
    angle = np.linspace( 0 , 2 * np.pi , 150 )
    x = a * np.cos( angle )
    y = b * np.sin( angle )
    plt.plot(x,y, linestyle=style, color=col)

def arrow(x1, y1, x2, y2, col="black"):
    plt.plot([x1, x2], [y1, y2], color=col)

def plot(th):
    ellipse(2*r, 2*r, "dashed", col="green")
    ellipse(r, r, col="orange")
    ellipse(a,b, col="blue")
    th = np.pi /180*th
    if(th==np.pi/2):
        alpha = np.pi/2
    else:
        alpha = np.arctan(b*np.tan(th)/a)
    l = np.sqrt(a*a*np.sin(alpha)*np.sin(alpha)+b*b*np.cos(alpha)*np.cos(alpha))
    if(l<2*r):
        d_angle = np.arccos(l/2/r)
    else:
        d_angle=0
    arrow(0,0, a*np.sin(alpha), b*np.cos(alpha), col="red")
    arrow(0,0, r*np.sin(th-d_angle), r*np.cos(th-d_angle), col="grey")
    arrow(r*np.sin(th-d_angle), r*np.cos(th-d_angle), r*np.sin(th-d_angle)+r*np.sin(d_angle+th), r*np.cos(th-d_angle)+r*np.cos(d_angle+th), col="black")
r = 0.625
a = 1
b = 2
lim = 2.5
arr = [0, 45, 60, 90]
plt.figure(figsize=(15,15), dpi=80)
plt.suptitle("Type-I Phase Matching")
for th in range(len(arr)):
    plt.subplot(2, 2, th+1)
    plot(arr[th])
    plt.xlim(-lim,lim)
    plt.ylim(-lim,lim)
    plt.gca().set_aspect('equal')
    plt.title("Angle = "+str(arr[th]))


# Show the plot
plt.show()
```

#SPDC in BBO crystal


In a practical laboratory setting, several considerations are crucial for efficiently generating Spontaneous Parametric Down-Conversion (SPDC) in an optical setup. Firstly, the phase-matching condition, where the change in wave vector ($\Delta k$) equals zero, must be as closely met as possible. It's challenging to find natural materials that naturally fulfill the condition where the refractive indices of the signal and idler waves ($n_i(\omega_i)$ and $n_s(\omega_s)$) are less than the refractive index of the pump wave ($n_p(\omega_p)$) and where the frequencies of the signal and idler waves ($\omega_i = \omega_s$) are lower than the pump frequency ($\omega_p$). However, this condition can be achieved using birefringent crystals, which have two (uniaxial crystals) or three (biaxial crystals) different indices for a given wavelength.

Another critical aspect is the desire for non-collinear phase-matching, leading to distinct wave vector directions for the three waves involved. This non-collinear configuration makes the theoretical aspects more complex but ensures the spatial separation of the twin photons.

For example, one of the most commonly used nonlinear crystals for SPDC is beta barium borate (BBO). BBO is a uniaxial crystal with a negative birefringence, meaning it has a fast axis with a low refractive index and a slow axis with a high refractive index. BBO is transparent across a wide range of wavelengths, from 190 nm to 3300 nm. The refractive indices of BBO are described by the Sellmeier equations. For type I down-conversion ($e \mapsto o + o$), the equations for the ordinary and extraordinary refractive indices ($n_o$ and $n_e$) are as follows:


$$n_o = 2.27405 + \frac{0.0184}{\lambda ^2 - 0.0179} - 0.0155 \lambda ^2$$
$$n_e = 2.3730 + \frac{0.0128}{\lambda^2 - 0.0156} - 0.0044 \lambda ^2$$

with $\lambda$ given in $\mu m$. We recall the relation of the extraordinary index of refraction with the angle $\theta$ from the optical axis is given by the ellipsoid equation for indices:

$$n^2(\theta) = \frac{n_e^2 n_o^2}{n_e^2 cos^2 \theta + n_o^2 sin^2 \theta}$$

So typically, when the pump beam enters the crystal with a certain angle $\theta$ with the optical axis OA,
the extraordinary index will be different for different angles, as opposed to the ordinary index. For
BBO, phase-matching conditions occur with an extraordinary beam only and one can have two types
of phase-matching:
$$e \longmapsto o + o : \text{Type I}$$
$$e \longmapsto e + o : \text{Type II}$$
$$e \longmapsto o + e : \text{Type II}$$

# Applications of SPDC

## Single Photon Source

SPDC results in the simultaneous creation of twin photons that are indistinguishable. This property is particularly valuable for low-signal level communications. Mandel and Rarity predicted that joint measurements can confirm the indistinguishability of these "twin" photons.

Hence, SPDC can be used to localize a one-photon state, leading to the development of heralded single photon sources. These sources exhibit strong photon antibunching, a distinct quantum property of light. While they are useful for quantum information processing (QIP) proofs of principle, true single photon sources like quantum dots or NV centers are preferred for practical applications.

The figure below illustrates an example of a heralded single photon source using SPDC. In this case, a collinear source of orthogonal twin photons is employed, and one photon (idler photon) is detected, which serves as a herald for the arrival of the signal photon. The signal photon is then detected using a Hanbury-Brown and Twiss (HBT) interferometer, a standard tool in quantum optics for characterizing single photons.

## Source of Indistinguishable Photons

In 1987, Leonard Mandel's group conducted a two-photon interferometry experiment. They generated a pair of twin photons through SPDC and then directed these photons to interfere on a beamsplitter, as depicted in Figure 5.

Quantum-mechanically, it was observed that when the twin photons exited the beamsplitter from different ports, they exhibited destructive interference, effectively canceling each other out. However, in cases where the photons emerged together from the beamsplitter, they exhibited a bunching effect, indicating a high probability of exiting the beamsplitter simultaneously. This phenomenon is the result of the bosonic nature of photons.

Bosons, like photons, follow the rules of quantum mechanics that allow multiple particles to occupy the same quantum state, unlike fermions such as electrons, which exhibit an aversion to being in the same state. The indistinguishability of the photons is crucial for this interference to occur. They must originate from the same optical mode, ensuring that their properties align, and this is the essence of the Hong-Ou-Mandel effect.

## Entangled Photons

Researchers employed parametric down-converted photons to create entangled photons in different polarization states, a crucial step in the study of quantum entanglement and Bell's inequalities. Using a type-II BBO crystal and a specific arrangement, they managed to produce two intersections of the output cones of twin photons with orthogonal polarizations. These photons were chosen because, under the right conditions, they become indistinguishable, making it impossible to predict the outcome of a polarization measurement at one of the intersections. However, it is known that parallel measurements on these photons would exhibit perfect anti-correlation in polarization.

In this entangled state, denoted as |ùúì‚ü©, the photons are correlated in a way that when one is measured in horizontal (H) polarization and the other in vertical (V) polarization, or vice versa, they will be found to be in opposite polarization states. The entangled state |ùúì‚ü© is represented as:

$$
\psi = \frac{1}{\sqrt{2}} (H_i V_s + e^{i\phi} V_i H_s)
$$

Where:

$H$ is for horizontal polarization, $V$ is for vertical polarization, $s$ and $i$ stand for signal and idler, $\phi$ is a phase that can be adjusted.

This entangled state allowed researchers to violate Bell's inequalities, a critical test of the principles of quantum mechanics. The experimental data illustrated strong correlations in polarization between the two entangled photons, challenging classical notions of physics. The entangled state and the techniques used to test Bell's inequalities sparked discussions related to the Einstein-Podolsky-Rosen (EPR) paradox, highlighting the peculiar and non-classical nature of quantum entanglement.


##SPDC in Quantum Technologies
Entanglement is one of the intriguing features of quantum mechanics and represents non-local features of a quantum state. Entanglement has numerous applications in quantum technologies like quantum computation, quantum communication, quantum networks and many more. Although the concept of entanglement has been explored in detailed in theory, its realisation is practice still remains one of the challenges. Previously we have seen that we can generate entangled photons using SPDC. Hence, one would like to use the source of entangled photons in quantum technologies. ###Quantum Teleportation
As the name suggest, this protocol allows us to "teleport" a quantum state. Though unlike teleportation portrayed in science fiction, quantum teleportation transfers only quantum information from one place to another. This protocol is very important for quantum communications. The sender does not need to know the particular state being transferred as well as the location of the recipient. However sender needs to send classical information to the recipient, and hence quantum teleportation cannot occur faster than the speed of light.

Now we will see how this protocol works. Suppose the sender (Alice) and the recipient (Bob) share an EPR pair given by $|\phi^{+}_{AB}\rangle=\frac{|0_a\rangle|0_B\rangle+|1_A\rangle|1_B\rangle}{\sqrt{2}}$ and have a classical communication channel between them. Alice then prepares a state she needs to teleport to Bob, given by $|\psi\rangle=\alpha|0\rangle+\beta|1\rangle$. Alice then performs Bell state measurement (BSM) on her qubits, and sends the measurement results she got to Bob via classical communication channel. Bob then performs single qubit operation on his qubit depending on the classical information he received from Alice. Finally he will end up with the state $|\psi\rangle$.

The complete protocol is shown below:

  <img src="https://drive.google.com/uc?export=view&id=1m2A-3S2TrLXCPFSNZizwisq0xV5vJ8l1" alt="Image Description">
</p>

The state evolution happens as follows:

1. Complete state of Alice (two qubits) and Bob (single qubit) is:
    $$|\psi_0\rangle=\frac{1}{\sqrt{2}}\left(\alpha|00_A\rangle|0_B\rangle+\beta|11_A\rangle|1_B\rangle\right)$$
  
2. After $CNOT$ gate is applied to Alice's qubit: $$|\psi_1\rangle=\frac{1}{\sqrt{2}}\left(\alpha|00_A\rangle|0_B\rangle+\beta|10_A\rangle|1_B\rangle\right)$$

3. After Hadamard gate is applied to first qubit:
    \begin{align}
        |\psi_2\rangle=&\frac{1}{2}[\alpha(|0_A\rangle+|1_A\rangle)(|0_a\rangle|0_B\rangle+|1_A\rangle|1_B\rangle)\\
        &+\beta(|0_A\rangle-|1_A\rangle)(|0_a\rangle|1_B\rangle+|1_A\rangle|0_B\rangle)]
    \end{align}

    which can be re-written as:

    \begin{align}
        |\psi_2\rangle=&\frac{1}{2}[|00_A\rangle(\alpha|0_B\rangle+\beta|1_B\rangle)+|01_A\rangle(\alpha|1_B\rangle+\beta|0_B\rangle)\\
        &+|10_A\rangle(\alpha|0_B\rangle-\beta|1_B\rangle)+|11_A\rangle(\alpha|1_B\rangle-\beta|0_B\rangle)]
    \end{align}

4. Finally Alice makes a joint measurement on two of her state and get the result $(m_1,m_2)$ where $m_1,m_2\in\{0,1\}$ and sends them to Bob who then makes control operation on his state: $Z^{m_1}X^{m_2}$

5. After the final operations by Bob, he ends up with the state $\alpha|0_B\rangle+\beta|1_B\rangle$

Now we will see how quantum teleportation was actually realised in 1997 at the University of Innsbruck.

1. Alice generates entangled photons 2 \& 3 using SPDC by passing a pulse of ultraviolet radiation through a nonlinear crystal and the photon 3 is sent to Bob

2. The retro-reflected beam creates another pair of photons, one of which will be prepared in the initial state of photon 1 to be teleported and the other one serving as a trigger indicating that a photon to be teleported is under way

3. In order to perform BSM, we need to make photons 1 \& 2 indistinguishable and hence we superimpose them on a beam splitter

4. Alice then performs BSM to project the state on a particular Bell state and sends that information to Bob over classical channel, who then ends up with the desired state

###Quantum Cryptography

Having seen quantum teleportation as a quantum communication protocol, now we look at another protocol, namely quantum cryptography, which is a very secure protocol and is based on use of entanglement and \textit{no-cloning theorem}. Here we will see how SPDC can be used to perform this protocol.

We use polarisation (up or down) of a photon to encode the information in a qubit. First the entangled photons are generated using SPDC and are transmitted via optical fibers to Alice and Bob, who are separated by some distance.

$$|\Psi^-\rangle=\frac{|0_A\rangle|1_B\rangle-|1_A\rangle|0_B\rangle}{\sqrt{2}}$$

Both of them then analye their photons independently by varying the measurement directions between $0$ and $\pi/4$.

###Quantum Computation
Quantum computing is an emerging field of computation where the computer uses *qubits* or *quantum bits* instead of nomral bits. Unlike normal bits, a qubit can be in any arbitrary superposition of two states: $\alpha|0\rangle+\beta|1\rangle$ with $\alpha,\beta$ being complex numbers and $|\alpha|^2+|\beta|^2=1$.

We have seen SPDC being used in quantum communication protocols in previous sections. As these advances were being developed, people also developed various experiments to use SPDC in quantum computation. The first step towards this was achieved by E. Knill, R. Laflamme and G. Milburn in 2001, who proposed a model which performs a $CNOT$ gate, which along with arbitrary qubit rotation gate, is part of universal set of gates for quantum computation. Here we provide the details of realisation of the $CNOT$ gate:

1. The control and target qubits are each encoded by a single photon across two spatial modes which is achieved by first generating pairs of degenerate photons using SPDC and passing them through polarising beam splitter (PBS) and half waveplate (HWP) and hence the polarisation encoded photonic qubits are converted into spatially encoded qubits
    
2. The two target modes $(|0\rangle$ and $|1\rangle)$ are mixed and recombined on two $50\%$ reflective beam splitters to form an interferometer which also includes a $33\%$ reflective beam splitter
    
3. The interferometer is balanced so that when the control photon is absent (state $|0\rangle$), the target qubit exits in the same state as it entered with
    
4. When control qubit is present (state $|1\rangle$), target photons interfere non-classically at the central $33\%-BS$ due to path indistinguishability which causes a phase shift of $\pi$ and hence the target state is output is flipped
