**<h1>Some key terms used<h1>**

**Geometric Phase** : It arises from the geomteric nature of the space a state inhabits. After a cycle of evolution, the state can acquire a  phase factor in addition to its usual dynamic phase. This is known as the geometric phase aquired.

**Dynamic Phase** : It is the phase acquired under the time integral of the Hamiltonian (thus the name "dynamic")

**Lorentz Reciprocity** : It is essentially the statement that the Greens Function in electromagnetism is complex-symmetric.


**<h1>Motivation<h1>**

**Why Geometric Phase?**
The breaking of Lorentz Reciprocity in optical systems is of great practical interest, mainly in the areas of optical isolators and on-chip optical systems.

Among these various non-reciprocal devices, a promising application  is in the area of controlable non-reciprocal phase shift.

An advantage of using geometric phase for this purpose is its *robustness* against parameter variation.

In the letter, a non-reciprocal manifestation of geometric phase in nonlinear frequency conversion has been introduced.

**<h1>Sum-Frequency-Generation<h1>**

The frequencies are denoted as  $ω_{1}$ : Pump frequency, $ω_{2}$ : Signal,
  $ω_{3}$ : Sum-frequency.

  So
   \begin{equation}
  \boxed{ω_{3} =  ω_{1} +  ω_{2}}
   \end{equation}

**<h2>Approximations used and Notation<h2>**

**<h4>Undepleted Pump Approximation<h4>**

Throughout the paper the undepleted pump approximation has been considered, that is the amplitude of the pump, $A_{1}(z)$ remains essentially constant through out the process.


**<h4>Slowly Varying Envelope Approximation<h4>**

Under this we can write the electric field of each wave as :
\begin{equation}
E_{n}(r,t) = A_{n}(z)E(x,y)e^{i(β_{n}z-ω_{n}t)} + c.c
\end{equation}

Where $A_{n}(z)$ is the slowly varying envelope and $β_{n}$ is the propagation constant of each wave.

For the rest of the paper the new amplitudes are defined as :
\begin{equation}
a_{i}(z) = \sqrt β_{i}  ω_{i}^{-1} A_{i}(z)
\end{equation}

So, in this notation the dynamics can be written as  :

\begin{equation}
i\frac{d}{dz} a_{2}(z) = κ^{*} e^{-iΔβz}a_{3}(z)
\\
i\frac{d}{dz} a_{3}(z) = κ e^{iΔβz}a_{2}(z)
\end{equation}
These can be written in a more compact matrix notation as :
\begin{equation}
i\frac{d}{dz}|\tilde{Ψ}⟩ =
\begin{pmatrix} -Δβ/2 & κ^{*}\\ κ & Δβ/2  \end{pmatrix}|\tilde{Ψ}⟩
\end{equation}

Here, $|\tilde{Ψ}⟩$ is the vector $\frac{[\tilde{a}_{2},\tilde{a}_{3}]^{T}}{\sqrt{|a_{2}|^{2}+|a_{3}|^{2}}}$, and the tilda amplitudes are the ones in the rotating frame.

**<h2>Geomteric Phase for SFG<h2>**

The geometric phase which is the Panchratnam-Berry phase can be defined here and can be obtained by removing the dynamic phase from the total phase accumulated.

**<h4>Calculation and Plot for Geometric Phase<h4>**
We start with the intial conditions $[1,0]^{T}$, for this intial condition the geometrical phase acquired is :



$γ(z) = \operatorname{arg}[\tilde{a}_{2}^{*}(0)\tilde{{a}}_{2}(z)] - Δβz/2$

which is exactly the total phase of $a_{2}(z)$.

Thus, the phase of $a_{2}$ is the pure geometric phase of $|\tilde{Ψ}⟩$, so using the analytical solution we get :

$ γ = arg{((cos(gz) + iΔβsin(gz)/2g)e^{-iΔβz/2}}) $

The plot for this is given below for various values of phase mismatch :


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the symbolic constants
d, c, b2, b3, A1, w2, w3 = sp.symbols('d c b2 b3 A1 w2 w3')

# Define the constant kappa symbolically
kappa = -2 * d * w2 * w3 / ((c**-2) * (b2 * b3)**(-1/2)) * A1

def h(x, a):
# h(x,a) = f(x, a)*g(x,a)
    real_part1 = np.cos(np.pi * np.sqrt(a**2 + 4) * x / 2)
    imag_part1 = a / np.sqrt(a**2 + 4) * np.sin(np.pi * np.sqrt(a**2 + 4) * x / 2)
    real_part2 = np.cos(np.pi*a*x)
    imag_part2 = - np.sin(np.pi*a*x)
    real_part3 = real_part1 * real_part2 - imag_part1 * imag_part2
    imag_part3 = real_part1 * imag_part2 + real_part2 * imag_part1
    return real_part3  +  1j * imag_part3

# Define the range of x values
x = np.linspace(0, 5, 500)

# Define a range of 'a' values
a_values = np.linspace(0, 3, 3)

# Plot the argument (phase) of f(x) for different 'a' values
for a in a_values:
    h_x = h(x, a)
    phase = np.angle(h_x)
    plt.plot(x, phase, label=f'a = {a:.2f}')

# Set plot labels and legend
plt.xlabel('x')
plt.ylabel('Argument (Phase) of h(x)')
plt.title('Plot of Argument of Berry Phase(x) vs. z|k|/2*pi with Different Values of phase mismatch')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

**<h1>Bloch Sphere representation<h1>**

We can visualize the state $|\tilde{Ψ}⟩$ by a Bloch Sphere, where the geometric phase is exactly half the solid angle with a surface area covered by the trajectory of the state vector on the sphere.

For cyclic evolutions, the geometric phase only depends on the solid angle that the trajectory encloses on the Poincaré sphere.

**<h3>Coordinates on the Bloch Sphere<h3>**

Given the state we can construct the density matrix :
\begin{equation}
ρ = |\tilde{Ψ}\rangle ⟨\tilde{Ψ}|
\end{equation}
Now, since this is a 2 by 2 matrix we can always expand it in the basis $(I,σ^{i})$, and we define the coordinates on the Bloch sphere by the components of $σ^{i}$, and to get them we can use the inner product :

\begin{equation}
S^{i} = Tr[ρσ^{i}]
\end{equation}

```python
import numpy as np
from qutip import Bloch, basis
import matplotlib.pyplot as plt

# Function to create a quantum state for a given z value
def quantum_state_vector(z, k, B_coeff):
    a = np.sqrt(4 * abs(k)**2 + B_coeff**2) / 2
    first_entry = np.cos(a * z) + 1j * (B_coeff / (2 * a)) * np.sin(a * z)
    second_entry = -1j * (k / a) * np.sin(a * z)
    return first_entry * basis(2, 0) + second_entry * basis(2, 1)

# Values for z and k
z_values = np.linspace(0, 2 * np.pi, 100)
k = 2  # You can change the value of k as needed
B_coeff = abs(k) * 100  # Adjust the coefficient as needed

# Initialize the Bloch Sphere
b = Bloch()

# Calculate and plot the state vector for each value of z
for z in z_values:
    psi = quantum_state_vector(z, k, B_coeff)
    b.add_states(psi)

# Show the Bloch Sphere
b.show()

plt.show()
```

The proposed method is the use of photonic transitions in a  nonlinear waveguide with a band structure.

A lithium niobate wave-guide has been employed with two fundamental modes of **horizontal** and **vertically** polarized electromagnetic waves.

The dependency of the frequency $\omega$ on the wavevector is given by :

Now, between these bands a transition can occur when momentum conservation and energy conservation are satisfied during the frequency conversion, clearly such  a transtion can occur  from the v band to h band as shown in the figure.


Suppose intially the pump is prepared in the v mode to drive the transitions between signal and the sum frequency in the h mode.

Now, we can have two cases :


```python
Image('fb.png',width = 400)
```

Now, the transition here is only permitted for the forward case, where $Δβ$ is almost 0, but seeing from our earlier graph of Geometrical Phase we know when phase mismatch is zero we pick up a geometrical phase of $\pi$ after each cycle.


For the backward case, the phase mismatch is very large, and from our graph of geometric phase we saw thar for large phase mismatch the geometrical phase is almost 0. A large phase mismatch in non-linear crystal can lead to self-focusing and spatial soliton formation in a crystal.

**NOTE :** Geometric phase often relies on the interference of different components of a wave. When waves with significantly different phase velocities interfere, they can create destructive interference.


Thus, we see a non - reciprocal manifestation of geometrical phase.
