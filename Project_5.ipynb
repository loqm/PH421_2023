#Density Matrix Equations of Motion for a Two-Level Atom

We are first considering the density matrix equations of motion for a two-level system without damping effects. This model doesn't have a unique way to include damping to the system. The interaction we are treating is below. The Hamiltonian for the system is:

\begin{equation}
    \hat{H} = \hat{H}_0 + \hat{V}(t) \tag{1.1}
\end{equation}
**Where:**
$\hat{H}_0$ is atomic Hamiltonian and
$\hat{V}$ interaction energy of atom with electromagnetic field

<p align="center">
  <img src="https://drive.google.com/uc?id=1qZcbMcsm5ahgPNVcCD7Vnk7xafL9_uxd" alt="Image Description">
</p>

<!-- ![picture](https://drive.google.com/uc?id=1qZcbMcsm5ahgPNVcCD7Vnk7xafL9_uxd#center) -->



\begin{equation}
    E_a = \hbar{\omega_a} \;\; and \; \; E_b = \hbar{\omega_b} \tag{1.2}
\end{equation}

The Hamiltonian $\hat{H}_0$ is a diagonal matrix whose elements are:
\begin{equation}
    \hat{H}_{0,nm} = E_n{\delta_{nm}} \tag{1.3}
\end{equation}

We are assuming that the interaction energy can be described in electric dipole approximation, hence
\begin{equation}
    \hat{V}(t) = -\hat{\mu}\bar{E}(t) \tag{1.4}
\end{equation}
The diagonal matrix elements of $\hat{\mu}$ are also zero, our assumption is that states a and b have definite parity. Thus $V_{aa} = V_{bb} = 0$ \\

The only non-zero elements of $\hat{V}$ are $V_{ab}$ and $V_{ba}$ which is
\begin{equation}
    V_{ba} = V^*_{ab} = -{\mu_{ba}}\bar{E}(t) \tag{1.5}
\end{equation}

The density matrix of the system is:
\begin{gather}
  \hat{\rho} =
  \begin{bmatrix}
    \rho_{aa} & \rho_{ab} \\
    \rho_{ba} & \rho_{bb} \\
  \end{bmatrix} \tag{1.6}
\end{gather}

**Where:** \\
$\rho_{ba}$ = $\rho^*_{ab}$ \\


The time evolution of the density matrix is:
\begin{equation}
    \dot{\rho}_{nm} = \frac{-i}{\hbar} [\hat{H}, \hat{\rho}]_{nm} = \frac{-i}{\hbar} [(\hat{H}\hat{\rho})_{nm} - (\hat{\rho}\hat{H})_{nm}] \tag{1.7}
\end{equation}

\begin{equation}
    = \frac{-i}{\hbar} {\sum\limits_{\nu}(H_{n\nu}\rho_{\nu m} - \rho_{n\nu}H_{\nu m})} \tag{1.8}
\end{equation}

\begin{equation}
    \dot{\rho}_{nm} = -i\omega_{nm}\rho_{nm} \frac{-i}{\hbar} {\sum\limits_{\nu}(V_{n\nu}\rho_{\nu m} - \rho_{n\nu}V_{\nu m})} \tag{1.9}
\end{equation}

Where:
$\omega_{nm} = \frac{(E_{n} - E_{m})}{\hbar}$

For the two-level atom, the indices n, m, and $\nu$ can take on the values a or b only. Thus the equations are:\\
\begin{equation}
    \dot{\rho}_{ba} = -i\omega_{ba}\rho_{ba} + \frac{i}{\hbar} {V_{ba}(\rho_{bb} - \rho_{aa})} \tag{1.10}
\end{equation}

\begin{equation}
    \dot{\rho}_{bb} = \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.11}
\end{equation}
\begin{equation}
    \dot{\rho}_{aa} = \frac{i}{\hbar} {(V_{ab}\rho_{ba} - \rho_{ab}V_{ba})} \tag{1.12}
\end{equation}

Adding equations 10 and 11 gives: \\
\begin{equation}
     \dot{\rho}_{bb} + \dot{\rho}_{aa} = 0 \tag{1.13}
\end{equation}
All the above equations neglect the presence of a relaxation process, an example case being when you supply a short duration pulse.

Next, we look at how these equations change in relaxation processes.


#Closed Two-Level Atom
Assume that upper level $b$ decays to lower level $a$ at a rate $\Gamma_{ba}$ or lifetime of upper level = $T_{1}$ = $\frac{1}{\Gamma_{ba}}$, typically a spontaneous emission

Called a **closed** system because any population leaving upper level surely goes to the lower level

Assumption is that atomic dipole moment is dephased with characteristic time $T_{2}$, causing transition linewidth (for weak applied fields) of width = $\gamma_{ba}$ = $\frac{1}{T_{2}}$

We can modify eqns (1.10), (1.11) and (1.12) as follows:
\begin{equation}
    \dot{\rho}_{ba} = -\left (i\omega_{ba} + \frac{1}{T_{2}}\right)\rho_{ba} + \frac{i}{\hbar} {V_{ba}(\rho_{bb} - \rho_{aa})} \tag{1.14}
\end{equation}
\begin{equation}
    \dot{\rho}_{bb} = -\frac{\rho_{bb}}{T_{1}} - \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.15}
\end{equation}
\begin{equation}
    \dot{\rho}_{aa} = \frac{\rho_{bb}}{T_{1}} + \frac{i}{\hbar} {(V_{ab}\rho_{ba} - \rho_{ab}V_{ba})} \tag{1.16}
\end{equation}

Hence,

\begin{equation}
    \dot{\rho}_{aa} + \dot{\rho}_{bb} = 0 \tag{1.17}
\end{equation} is still satisfied.


Closely observing Eqn (1.14) tells us that it depends on only on the difference of the populations ($\rho_{aa}$ - $\rho_{bb}$)

Subtracting (1.16) from (1.15) gives,

\begin{equation}
    \frac{d}{dt}\left(\rho_{bb} - \rho_{aa}\right) = -\frac{2\rho_{bb}}{T_{1}} - \frac{2i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.18}
\end{equation}

Writing $2\rho_{bb}$ = ($\rho_{bb}$ - $\rho_{aa}$) + ($\rho_{bb}$ + $\rho_{aa}$), equation (17) becomes
\begin{equation}
    \frac{d}{dt}\left(\rho_{bb} - \rho_{aa}\right) = -\frac{\rho_{bb} - \rho_{aa}+1}{T_{1}} - \frac{2i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.19}
\end{equation}
given that $\rho_{bb}$ + $\rho_{aa}$ = 1
For a general relation, considering some \newline ($\rho_{bb}$ - $\rho_{aa}$)$^{(eq)}$ in thermal equilibrium and replacing -1 in Eqn (18) which assumes that in thermal equilibrium all population resides in ground state, we can write it as:
\begin{equation}
    \frac{d}{dt}\left(\rho_{bb} - \rho_{aa}\right) = -\frac{(\rho_{bb} - \rho_{aa}) - (\rho_{bb} - \rho_{aa})^{(eq)}}{T_{1}} - \frac{2i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.20}
\end{equation}

\begin{frame}{Closed Two-Level Atom}
Let us examine the solution the solutions to these equations in the absence of an applied field, i.e $V_{ba}$ = 0

\begin{equation}
    \rho_{bb}(t) - \rho_{aa}(t) = (\rho_{bb} - \rho_{aa})^{(eq)}) + \left[(\rho_{bb}(0) - \rho_{aa}(0)) - (\rho_{bb} - \rho_{aa})^{(eq)})\right]e^{\frac{t}{T_{1}}} \tag{1.21}
\end{equation}

Similarly, the solution for Eqn (13) will be
\begin{equation}
    \rho_{ba}(t) = \rho_{ba}(0)e^\frac{-(i\omega_{ba} + 1)t}{T_{2}} \tag{1.22}
\end{equation}

This can also be interpreted as the expectation value of the induced dipole moment, given by:

\begin{equation}
    \left\langle \tilde{\mu}(t)\right\rangle = Tr(\hat{\rho}\hat{\mu}) = \mu_{ab}\rho_{ba}(t) + \mu_{ba}\rho_{ab}(t) = \mu_{ab}\rho_{ba}(0)e^{-\frac{(i\omega_{ba}+1)t}{T_{2}}} + c.c  \tag{1.23}
\end{equation}


\begin{equation*}
    = \left[\mu_{ab}\rho_{ba}(0)e^{-i\omega_{ba}t} + c.c\right] e^{\frac{t}{T_{2}}} \tag{1.24}
\end{equation*}

This result shows that for undriven atom, dipole moment oscillates at frequency $\omega_{ba}$ and decays to zero in time $T_{2}$ also known as dipole dephasing time.

Due to ..., $T_{1}$ and $T{2}$ are related to the collisonal dephasing rate ($\gamma_{c}$) by
\begin{equation}
    \frac{1}{T_{2}} = \frac{1}{2T_{1}} + \gamma_{c} \tag{1.25}
\end{equation}

For atomic vapour, $\gamma_{c}$ is described by the formula
\begin{equation}
    \gamma_{c} = C_{s}N + C_{f}N_{f} \tag{1.26}
\end{equation}
where $N$ = No. density of atoms having resonance frequency $\omega_{ba}$,$N_{f}$ = No. density of \textit{foreign} atoms of different atomic species with different resonant frequency $C_{s}$, $C_{f}$ = Coefficients describing self-broadening \& foreign-gas broadening respectively\\

As compared to the closed system, the open two-level atom allows the exchange of population with associated reservoir levels (like magnetic sublevels, hyperfine levels associated with states a and b) as seen in the figure above.

For reasons that the population leaving the upper level does not necessarily end up to the lower level, these systems are called \textbf{open} systems.


These types of systems are observed in laser theory, where both upper and lower levels are assumed to acquire population at controllable pump rates, namely $\lambda_{a}$ and $\lambda_{b}$. The density matrix equations now are:
\begin{equation}
    \dot{\rho}_{ba} = -\left (i\omega_{ba} + \frac{1}{T_{2}}\right)\rho_{ba} + \frac{i}{\hbar} {V_{ba}(\rho_{bb} - \rho_{aa})} \tag{1.27}
\end{equation}
\begin{equation}
    \dot{\rho}_{bb} = \lambda_{b} -\Gamma_{b}(\rho_{bb} - \rho{bb}^{(eq)}) - \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.28}
\end{equation}
\begin{equation}
    \dot{\rho}_{aa} = \lambda_{a} -\Gamma_{a}(\rho_{aa} - \rho_{aa}^{(eq)})+ \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.29}
\end{equation}
Note that here the total population in the two levels is not conserved i.e
$\dot{\rho}_{aa}$ + $\dot{\rho}_{bb}$ $\neq$ 0 or $\rho_{aa}$ + $\rho_{bb}$ $\neq$ 1

The relaxation rates are related to the collisional dephasing rate $\gamma_{c}$ and population rates $\Gamma_{a}$ and $\Gamma_{b}$ by
\begin{equation}
    \frac{1}{T_{2}} = \frac{1}{2}(\Gamma_{a} + \Gamma_{b}) + \gamma_{c} \tag{1.30}
\end{equation}

This kind of system is often used to model a saturable absorber.

 Population leaves the optically excited level $b$ at a rate $\Gamma_{ba} + \Gamma_{bc}$
 Level $c$ acts as a trap level; population decays from level $c$ back to the ground state at a rate $\Gamma_{ca}$
Dipole moments associated with transition between levels $a$ and $b$ is damped at the rate $\gamma_{ba}$



With this, Eqns (1.14), (1.15) and (1.16) become,
\begin{equation}
    \dot{\rho}_{ba} = -\left (i\omega_{ba} + \gamma_{ba}\right)\rho_{ba} + \frac{i}{\hbar} {V_{ba}(\rho_{bb} - \rho_{aa})} \tag{1.31}
\end{equation}
\begin{equation}
    \dot{\rho}_{bb} = -(\Gamma_{ba} + \Gamma_{bc})\rho_{bb} - \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.32}
\end{equation}
\begin{equation}
    \dot{\rho}_{cc} = \Gamma_{bc}\rho_{bb} - \Gamma_{ca}\rho_{cc} \tag{1.33}
\end{equation}
\begin{equation}
    \dot{\rho}_{aa} = \Gamma_{ba}\rho_{bb} + \Gamma_{ca}\rho_{cc} + \frac{i}{\hbar} {(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})} \tag{1.34}
\end{equation}

Again by inspection, the population in the three levels is conserved, as
\begin{equation}
    \dot{\rho}_{aa} + \dot{\rho}_{bb} + \dot{\rho}_{cc} = 0 \tag{1.35}
\end{equation}


# Steady-State Response of a Two-Level Atom to a Monochromatic Field

We examine the nature of the solution to the density matrix equations of motion for a two-level atom in the presence of a monochromatic, steady-state field. We treat the case of a closed two-level atom. For the closed two-level atomic system, the density matrix equations were given as
\begin{equation}
    \frac{d}{dt}\rho_{ba} = -\left( i\omega_{ba} + \frac{1}{T_2}\right)\rho_{ba} + \frac{i}{\hbar} V_{ba} (\rho_{bb} - \rho_{aa})\tag{2.1}
\end{equation}
\begin{equation}
    \frac{d}{dt}(\rho_{bb} - \rho_{aa}) = -\frac{(\rho_{bb} - \rho_{aa}) - (\rho_{bb} - \rho_{aa})^{(eq)}}{T_1} - \frac{2i}{\hbar}(V_{ba}\rho_{ab} - \rho_{ba}V_{ab})\tag{2.2}
\end{equation}


The interaction Hamiltonian for an applied field in the form of a monochromatic wave of frequency $\omega$ is given by
\begin{equation}
    \hat{V} = -\hat{\mu}\bar{E}(t) = -\hat{\mu}(E e^{-i\omega t}+ E^* e^{i\omega t})\tag{2.3}
\end{equation}
and the matrix elements of the interaction Hamiltonian are then given by
\begin{equation}
    V_{ba} = -\mu_{ba}(E e^{-i\omega t}+ E^* e^{i\omega t})\tag{2.4}
\end{equation}
The density matrix equations cannot be solved exactly for $V_{ba}$ given by eq. (2.4). However, they can be solved in an approximation known as the rotating-wave approximation.

From eq. (2.1), we can verify that in the absence of a driving field, $\rho_{ba}$ tends to evolve in time as $e^{(-i\omega_{ba}t)}$. For this reason, when $\omega$ is approximately equal to $\omega_{ba}$, the part of $V_{ba}$ that oscillates as $e^{(-i\omega t)}$ acts as a far more effective driving term for $\rho_{ba}$ than does the part that oscillates as $e^{(i\omega t)}$. It is thus a good approximation to take $V_{ba}$ not as eq. (2.4) but instead as $V_{ba} = -\mu_{ba} E e^{-i\omega t}$.

This approximation is called the **rotating-wave approximation**. Within this approximation, the density matrix equations of motion become
\begin{equation}
    \frac{d}{dt}\rho_{ba} = -\left(i\omega_{ba} + \frac{1}{T_2}\right)\rho_{ba} + \frac{i}{\hbar} \mu_{ba}E e^{-i\omega t}(\rho_{bb} - \rho_{aa})\tag{2.5}
\end{equation}
\begin{align}
    \frac{d}{dt}(\rho_{bb} - \rho_{aa}) =& -\frac{(\rho_{bb} - \rho_{aa}) - (\rho_{bb} - \rho_{aa})^{(eq)}}{T_1}\tag{2.6}\\
    &-\frac{2i}{\hbar}(\mu_{ba}E e^{-i\omega t}\rho_{ab}-\mu_{ab}E^* e^{i\omega t}\rho_{ba})
\end{align}

We want to find the steady-state solution to these equations. We, therefore, introduce a slowly varying quantity $\rho_{ba} = \sigma_{ba}(t) e^{-i\omega t}$.

So now, equations eq. (2.5) and eq. (2.6) becomes,
\begin{equation}
    \frac{d}{dt} \sigma_{ba} = \left[ i(\omega - \omega_{ba}) - \frac{1}{T_2} \right]\sigma_{ba} - \frac{i}{\hbar} \mu_{ba} E (\rho_{bb} - \rho_{aa})\tag{2.7}
\end{equation}
\begin{align}
    \frac{d}{dt}(\rho_{bb} - \rho_{aa}) =& -\frac{(\rho_{bb} - \rho_{aa}) - (\rho_{bb} - \rho_{aa})^{(eq)}}{T_1}\tag{2.8}\\
    &-\frac{2i}{\hbar}(\mu_{ba}E e^{-i\omega t}\sigma_{ab}-\mu_{ab}E^* e^{i\omega t}\sigma_{ba})
\end{align}
For steady state solution, we set the LHS of these equations as zero and we get
\begin{align}
    \rho_{bb} - \rho_{aa} =& \; \frac{(\rho_{bb} - \rho_{aa})^{(eq)}[1 + (\omega - \omega_{ba})^2 T^2_2]}{1 + (\omega - \omega_{ba})^2 T^2_2 + (4/\hbar^2) |\mu_{ba}|^2 |E|^2 T_1 T_2} \tag{2.9}\\
    \rho_{ba} =& \; \sigma_{ba}e^{-i\omega t} = \frac{\mu_{ba} E e^{-i\omega t} (\rho_{bb} - \rho_{aa})}{\hbar (\omega - \omega_{ba} + i/T_2)} \tag{2.10}
\end{align}

Polarization, which is given in terms of the off-diagonal elements of the
density matrix by is [*(For proof of this result, refer)*](https://galileo.phys.virginia.edu/classes/752.mf1i.spring03/DensityMatrix.htm)
\begin{equation}
    \bar{P}(t) = N \langle \bar{\mu} \rangle = N \text{Tr}(\hat{\bar{\rho}} \hat{\bar{\mu}}) = N(\mu_{ab}\rho_{ba} + \mu_{ba}\rho_{ab})\tag{2.11}
\end{equation}
where $N$ is the number density of atoms. We define the complex amplitude P of the polarization and susceptibility $\chi$ as
\begin{equation}
    \bar{P}(t) = P e^{-i\omega t} + \text{c.c.} ; \;\;\;\;\;\;\;\; P = \epsilon_0 \chi E \tag{2.12}
\end{equation}
Substituting in this from eq. (2.10) and eq. (2.9), we get
\begin{align}
    \chi =& \frac{N(\rho_{bb} - \rho_{aa})^{(eq)}|\mu_{ba}|^2 (\omega - \omega_{ba} - i/T_2)T_2^2/\epsilon_0 \hbar}{1 + (\omega - \omega_{ba})^2 T_2^2 + (4/\hbar^2)|\mu_{ba}|^2 |E|^2 T_1 T_2} \\
    =& \left[ N(\rho_{bb} - \rho_{aa})^{(eq)}|\mu_{ba}|^2 \frac{T_2}{\epsilon_0 \hbar} \right] \frac{\Delta T_2 - i}{1 + \Delta^2 T_2^2 + \Omega^2 T_1 T_2}\tag{2.13}
\end{align}
where the on-resonance Rabi frequency and detuning frequency are $\Omega = 2|\mu_{ba}| |E|/\hbar$ and $\Delta = \omega - \omega_{ba}$ respectively.

Note that this expression gives the total susceptibility, including both its linear and nonlinear contributions.

The absorption coefficient is given in general by
\begin{equation}
    \alpha = \frac{2\omega}{c} \text{Im} n = \frac{2\omega}{c} \text{Im}\left[(1+\chi)^{1/2})\right] \tag{2.14}
\end{equation}
and, whenever the condition $|\chi| \ll 1$ is valid, the absorption coefficient can be expressed by $\alpha = (\omega/c) \text{Im} \chi$.

The term in square bracket can be redefined to simplify the expression for $\chi$ as,
\begin{equation}
    \chi = -\frac{\alpha_0(0)}{\omega_{ba}/c} \frac{\Delta T_2 - i}{1 + \Delta^2 T_2^2 + \Omega^2 T_1 T_2}\tag{2.15}
\end{equation}

```python
#Graph of Chi
# printing the graphs of the Real and imaginary parts of the susceptibility plotted as functions of the optical frequency ω for several values of the saturation parameter

import matplotlib.pyplot as plt
import numpy as np
import math

def chi_real_fun(x, saturation_parameter):
  return (1/math.sqrt(1 + saturation_parameter)) * (x/math.sqrt(1 + saturation_parameter)) / (1 + (x**2)*(1 + saturation_parameter))

def chi_imag_fun(x, saturation_parameter):
  return (1/(1 + saturation_parameter)) / (1 + (x**2)*(1 + saturation_parameter))

x = np.arange(-10, 10.1, 0.1)
saturation_parameters = [2, 4, 0.01]

figure, axis = plt.subplots(2,1)
for saturation_param in saturation_parameters:
  chi_real = chi_real_fun(x, saturation_param)
  chi_imag = chi_imag_fun(x, saturation_param)
  axis[0].plot(x, chi_real, label='$\\Omega^2 T_1 T_2$ = ' + str(saturation_param))
  axis[1].plot(x, chi_imag, label='$\\Omega^2 T_1 T_2$ = ' + str(saturation_param))

axis[0].legend()
axis[1].legend()
plt.xlabel('$\\Delta T_2$')
axis[0].set_ylabel('$Re(\\chi)$')
axis[1].set_ylabel('$Im(\\chi)$')
plt.show()
```

It is convenient to define, by means of the relation
\begin{equation}
    \Omega^2 T_1 T_2 = \frac{|E|^2}{|E_S^0|^2}; \;\;\;\;\;\;\;\; |E_S^0|^2 = \frac{\hbar^2}{4|\mu_{ba}|^2 T_1 T_2}\tag{2.16}
\end{equation}
the quantity $|E_S^0|$ is known as the line-center saturation field strength. The first and third-order contributions to the susceptibility can be found by performing a power series expansion of $|E|^2/|E_S^0|^2$ and retaining only the first and second
terms,
\begin{equation}
    \chi \simeq -\frac{\alpha_0(0)}{\omega_{ba}/c} \left( \frac{\Delta T_2 - i}{1 + \Delta^2 T_2^2} \right) \left( 1 - \frac{1}{1 + \Delta^2 T_2^2} \frac{|E|^2}{|E_S^0|^2}\right)\tag{2.17}
\end{equation}

We now equate this expression with the usual power series expansion $\chi = \chi^{(1)} + 3\chi^{(3)}|E^2|$ (where $\chi^{(3)} \equiv \chi^{(3)}(\omega = \omega + \omega - \omega$)) to find that the first and third-order susceptibilities are given by
\begin{align}
    \chi^{(1)} =& -\frac{\alpha_0(0)}{\omega_{ba}/c} \frac{\Delta T_2 - i}{1 + \Delta^2 T_2^2} \tag{2.18} \\
    \chi^{(3)} =& \frac{\alpha_0(0)}{3\omega_{ba}/c} \left[ \frac{\Delta T_2 - i}{(1 + \Delta^2 T_2^2)^2} \right] \frac{1}{|E_S^0|^2} \tag{2.19} \\
    =& -\frac{4}{3}N(\rho_{bb} - \rho_{aa})^{(eq)}|\mu_{ba}|^4 \frac{T_1 T_2^2}{\epsilon_0 \hbar^3} \frac{\Delta T_2 - i}{(1 + \Delta^2 T_2^2)^2} \tag{2.20}
\end{align}
In the limit $\Delta T_2 \gg 1$, the expression for $\chi^{(3)}$ reduces to
\begin{equation}\tag{2.21}
    \chi^{(3)} = -\frac{4}{3}N(\rho_{bb} - \rho_{aa})^{(eq)}|\mu_{ba}|^4 \frac{1}{\hbar^3 \Delta^3} \frac{T_1}{T_2}
\end{equation}

Let us consider the magnitudes of some of the physical quantities we
have been introduced in this section. Since (for $n = 1$) the intensity of an optical wave with field strength $E$ is given by $I = 2\epsilon_0 c|E|^2$, the Rabi frequency can be expressed as
\begin{equation}
    \Omega = \frac{2|\mu_{ba}|}{\hbar} \left( \frac{I}{2\epsilon_0 c}\right)^{1/2}\tag{2.22}
\end{equation}
Assuming that $|\mu_{ba}|= 2.5ea_0 = 2.0 \times 10^{-29}$ Cm (as it is for the $3s to 3p$ transition of atomic sodium) and that $I$ is measured in $W/cm^2$, this relationship gives the numerical result
\begin{equation}\tag{2.23}
    \Omega[\text{rad/sec}] = 2\pi (1 \times 10^9) \left(\frac{I[W/cm^2]}{127}\right)^{1/2}
\end{equation}

```python
#Graph of Chi3
# printing the graphs of the Real and imaginary parts of the third order susceptibility plotted as functions of the optical frequency ω for several values of the saturation parameter

import matplotlib.pyplot as plt
import numpy as np
import math

def chi_real_fun(x):
  return x/((1 + x**2)**2)

def chi_imag_fun(x):
  return -1/((1 + x**2)**2)

x = np.arange(-10, 10.1, 0.1)

figure, axis = plt.subplots(2,1)
chi_real = chi_real_fun(x)
chi_imag = chi_imag_fun(x)
axis[0].plot(x, chi_real, label='Real part of $\\chi^{(3)}$')
axis[1].plot(x, chi_imag, label='Imaginary part of $\\chi^{(3)}$')

axis[0].legend()
axis[1].legend()
plt.xlabel('$\\Delta T_2$')
axis[0].set_ylabel('$Re(\\chi^{(3)})$')
axis[1].set_ylabel('$Im(\\chi^{(3)})$')
plt.show()
```

Hence, whenever the intensity $I$ exceeds $127 W/cm^2$, $\Omega/2\pi$ becomes greater than 1 GHz, which is a typical value of the [Doppler-broadened linewidth of an atomic transition](https://en.wikipedia.org/wiki/Doppler_broadening).

Let us consider the magnitude of $\chi^{(3)}$ under conditions of the near-resonant excitation of an atomic transition. We take the typical values $N = 10^{14} cm^{-3}$, $(\rho_{bb} - \rho_{aa})^{(eq)} = -1$, $\mu_{ba} = 2.0 \times 10^{-29} Cm$, $\Delta = \omega - \omega_{ba} = 2\pi c (cm^{-1}) = 6\pi \times 10^{10} rad/sec$, and $T_2/T_1 = 2$, in which case we find from equation eq. (2.21) that $\chi^{(3)} = 2.1 \times 10^{-16} m^2/V^2$.

#Optical Bloch Equations

We can simplify the form using the following quantities:
1. The population inversions
\begin{equation}
  \omega =   \rho _{bb}-\rho _{aa} \: \: and \: \:  \omega ^{(eq)} =(\rho _{bb}-\rho _{aa})^{eq} \tag{3.1}
\end{equation}
2. The detuning of the optical field from resonance,
\begin{equation}
   \Delta = \omega -  \omega _{ba} \tag{3.2}
\end{equation}
3. The atom-field coupling constant
\begin{equation}
   k= \frac{2  \mu _{ba} }{\hbar} \tag{3.3}
\end{equation}
Now equations look like this:
\begin{equation}
  \frac{d}{dt} \sigma_{ba} = \left[ i\Delta - \frac{1}{T_2} \right]\sigma - \frac{ik}{2} E\omega \tag{3.4}
\end{equation}

\begin{align}
    \frac{d}{dt}(\omega) =& -\frac{\omega -\omega ^{(eq)}}{T_1}\\
    & -i(kE\sigma^*-k^*E^*\sigma)\tag{3.5}
\end{align}

It is useful to consider the equation of motion satisfied by the complex amplitude of the induced dipole moment. We know that the expectation value of the induced dipole moment is given by
\begin{equation}
   <    \overrightarrow{ \widetilde{ \mu } }  > =  \rho _{ba}  \mu _{ab}   + \rho _{ab}  \mu _{ba}=  \sigma_{ba}  \mu _{ab}  e^{-i \omega t}   +  \sigma _{ab}  \mu _{ba}e^{i \omega t} \tag{3.6}
\end{equation}
\begin{equation}
 <    \overrightarrow{ \widetilde{ \mu } }  >=pe^{-i \omega t} + c.c.\tag{3.7}
\end{equation}
\begin{equation}
     p=\sigma_{ba}  \mu _{ab} \tag{3.8}
\end{equation}
We can rewrite equation (3.4) \& (3.5) in terms of the dipole amplitude p
\begin{equation}
  \frac{d}{dt} \sigma_{ba} = \left[ i\Delta - \frac{1}{T_2} \right]p - \hbar|{k}|^2\frac{i}{4} E\omega \tag{3.9}
\end{equation}

\begin{align}
 \frac{d}{dt}(\omega) =& -\frac{\omega -\omega ^{(eq)}}{T_1}\\
    & - \frac{4}{\hbar}Im(Ep^*)\tag{3.10}
  \end{align}

Equations are linear in the atomic variables p and w and in the applied field amplitude E. However, the coupling is parametric
Let's look at the cases where $E$ is real in those cases we assume that the phase convention for describing the atomic energy eigenstates has been chosen such that $\mu _{ba} $ and hence $κ$ are real quantities.
In these cases, we can write $\sigma$ as:
\begin{equation}
     \sigma=1/2 (u-iv) \tag{3.11}
\end{equation}
So using this in equation (3.4) we get,
\begin{equation}
  \frac{d}{dt}(u-iv) = \left[ i\Delta - \frac{1}{T_2} \right](u-iv) -ikE\omega \tag{3.12}
\end{equation}
We can separate real and imaginary parts, then we will get
\begin{equation}
  \frac{d}{dt}u=\Delta v-u/ T_{2} \tag{3.13}
\end{equation}
\begin{equation}
  \frac{d}{dt}v=-\Delta u-v/ T_{2} +kE\omega\tag{3.14}
\end{equation}
and equation (3.5) becomes,
\begin{align}
 \frac{d}{dt}(\omega) =& -\frac{\omega -\omega ^{(eq)}}{T_1}
    - kEv \tag{3.15}
\end{align}
the set of equations given above are called optical Bloch equations
Now we will try to prove that in the absence of relaxation processes($T_1, T_2 \longrightarrow \infty$) the variables obey the conservation law;
\begin{equation}
  u^2+v^2+\omega ^2=1 \tag{3.16}
\end{equation}
for that let's first take the derivative of it and we can easily see that it goes to 0.
Let's note one thing before the optical field is applied the atom must be in its ground state and hence $w = −1$ and $u = v = 0$ and therefore $u^2+v^2+\omega ^2=1$.
Also, all the damping terms in Bloch equations are of negative sign therefore we can say that,
\begin{equation}
  u^2+v^2+\omega ^2<=1 \tag{3.17}
\end{equation}
#Harmonic Oscillator Form of the Density Matrix Equations
We can obtain different intuitions by considering the equation of motion satisfied by the expectation value of the dipole moment induced by the applied field. Let that be $\widetilde{M}$
\begin{equation}
  \widetilde{M} = <    \overrightarrow{ \widetilde{ \mu } }  > =  \rho _{ba}  \mu _{ab} +c.c. \tag{3.18}
\end{equation}
Now we will consider the following form of density matrix equations
\begin{equation}
  \frac{d}{dt}\rho_{ba} = -\left[ i\omega_{ba}+\frac{1}{T_2} \right]\rho_{ba} -\frac{i}{\hbar}\mu_{ba} \widetilde{E}\omega \tag{3.19}
\end{equation}

\begin{equation}
  \frac{d}{dt}\omega =-\frac{\omega-\omega^{(eq)}}{T_{1}}+\frac{4\widetilde{E}}{\hbar}Im(\mu_{ab}\rho_{ba}) \tag{3.20}
\end{equation}
It follows from equation (1.5), (1.10) \& $\omega = (\rho_{bb} - \rho_{aa})$
Let's take the first and the second time derivative of $\widetilde{M}$. Then we will get,     
\begin{equation}
  \frac{d}{dt}\widetilde{M}=\frac{d}{dt}(\rho _{ba})  \mu _{ab}+c.c.=-\left[ i\omega_{ba}+\frac{1}{T_2} \right]\rho_{ba}\mu_{ab}+c.c. \tag{3.21}
\end{equation}
\begin{equation}
  \frac{d^2}{dt^2}\widetilde{M}=(-\omega_{ba}^2+\frac{2i\omega_{ba}}{T_{2}}+\frac{1}{T_{2}^2})\rho_{ba}\mu{ab}-\frac{\omega_{ba}}{\hbar}\widetilde{E}\omega+c.c.\tag{3.22}
\end{equation}
Now if we substitute equations (3.18) and (3.21) in (3.22) we will get,
\begin{equation}
  \frac{d^2}{dt^2}\widetilde{M}+\frac{2}{T_{2}}\frac{d}{dt}\widetilde{M}+\omega_{ba}^2 \widetilde{M}=-\frac{\widetilde{M}}{T_{2}^2}-\frac{2\omega_{ba}}{\hbar}|\mu_{ba}|^2\widetilde{E}\omega \tag{3.23}
\end{equation}
We can make another realistic approximation because $\omega_{ba}^2$ is much larger than $1/T_2^2$
\begin{equation}
  \frac{d^2}{dt^2}\widetilde{M}+\frac{2}{T_{2}}\frac{d}{dt}\widetilde{M}+\omega_{ba}^2 \widetilde{M}=-\frac{2\omega_{ba}}{\hbar}|\mu_{ba}|^2\widetilde{E}\omega \tag{3.24}
\end{equation}
To simplify equation (3.20) we need to calculate $Im(\mu_{ab}\rho_{ba})$ for that we can rewrite equation (3.21) as:
\begin{equation}
  \frac{d}{dt}\widetilde{M}=-(i\omega_{ba}+\frac{1}{T_{2}})\rho_{ba}\mu_{ab}+c.c.=2\omega_{ba}Im(\rho_{ba}\mu_{ab})-\widetilde{M}/T_{2} \tag{3.25}
\end{equation}
which shows that
\begin{equation}
  Im(\rho_{ba}\mu_{ab})=\frac{1}{2\omega_{ba}}(\frac{d}{dt}\widetilde{M}+\frac{\widetilde{M}}{T_{2}}) \tag{3.26}
\end{equation}
Substitute equation (3.26) in (3.20)
\begin{equation}
 \frac{d}{dt}(\omega) = -\frac{\omega -\omega ^{(eq)}}{T_1}+ \frac{2\widetilde{E}}{\hbar\omega_{ba}}(\frac{d}{dt}\widetilde{M}+\frac{\widetilde{M}}{T_{2}}) \tag{3.27}
\end{equation}
Since $\dot{\widetilde{M}}$ oscillates at an optical frequency (which is much larger than 1/T2),
the term $\widetilde{M}/T_2$ can be omitted
\begin{equation}
 \frac{d}{dt}(\omega) = -\frac{\omega -\omega ^{(eq)}}{T_{1}} + \frac{2\widetilde{E}}{\hbar\omega_{ba}}(\frac{d}{dt}\widetilde{M}) \tag{3.28}
\end{equation}
#Adiabatic Following Limit
In this approximation, the central idea is no relaxation occurs during the optical pulse. So, We assume that the optical field is in the form of a pulse whose length $\tau_p$ obeys the following condition
\begin{equation}
     \tau _{p}  \ll  T_{1} ,T_{2}; \tag{3.29}
\end{equation}
Additionally, we assume that the laser is detuned sufficiently far from resonance
\begin{equation}
     |\omega-\omega_{ba}|\gg T_{2}^{-1},\tau_{p}^{-1},\mu_{ba}E/\hbar \tag{3.30}
\end{equation}
To simplify the following analysis, we introduce the Rabi frequency
\begin{equation}
      \Omega(t)=2\mu_{ba}E(t)/\hbar \tag{3.31}
\end{equation}
Let's write a density matrix equation in the limit of $T_1, T_2 \rightarrow \infty$
\begin{equation}
      \frac{d}{dt}\sigma=i\Delta\sigma-i\Omega\omega/2 \tag{3.32}
\end{equation}
\begin{equation}
      \frac{d}{dt}\omega=-i(\Omega^{*}\sigma-\Omega\sigma^{*}) \tag{3.33}
\end{equation}
Here we can do a similar analysis as we did in the case of equation (3.16) and we can conclude the following result
\begin{equation}
      \omega^{2}(t)+4|\sigma(t)|^{2}=1 \tag{3.34}
\end{equation}
We now make the adiabatic following approximation, that is, we assume that for all times the atomic response is nearly in a steady state with the applied field, therefore, ${d\sigma}/{dt}, {d\omega}/{dt}$ are zero. From this we can obtain the following relation.
\begin{equation}
      \sigma(t)=\omega(t)\Omega(t)/2\Delta \tag{3.35}
\end{equation}
Since $\omega(t)$ is a real quantity, $\sigma(t)$ \& $\Omega$ have same phase. Let's substitute equation (3.35) in (3.34)
\begin{equation}
    \omega(t)^{2}+\frac{\omega(t)^{2}|\Omega(t)|^{2}}{\Delta^{2}}=1 \tag{3.36}
\end{equation}
\begin{equation}
    \omega(t)=-\frac{|\Delta|}{(\Delta^{2}+|\Omega(t)|^{2})^{1/2}} \tag{3.37}
\end{equation}
\begin{equation}
    \sigma(t)=-\frac{1/2\Omega(t)}{|\Delta|}\frac{|\Delta|}{(\Delta^{2}+|\Omega(t)|^{2})^{1/2}} \tag{3.38}
\end{equation}
Using these results we can deduce the value of the nonlinear susceptibility
\begin{equation}
    P=N\mu_{ab}\sigma \tag{3.39}
\end{equation}
\begin{equation}
    P=-\frac{1/2\Omega(t)}{|\Delta|}\frac{N\mu_{ab}|\Delta|}{(\Delta^{2}+|\Omega(t)|^{2})^{1/2}} \tag{3.40}
\end{equation}
$|\Delta|>>|\Omega|$ is valid so we can expand the equation in the power series
\begin{equation}
    P=-\frac{1/2\Omega(t)}{|\Delta|^{2}}\frac{N\mu_{ab}}{(1+|\Omega|^{2}/\Delta^{2})^{1/2}}=\frac{1/2\Omega(t)}{|\Delta|^{2}} N\mu_{ab}(1-1/2 |\Omega|^{2}/\Delta^{2}) \tag{3.41}
\end{equation}
The third-order contribution of P will be as follows and it is $3\epsilon_{0}\chi^{(3)}|E|^2E$
\begin{equation}
    \frac{|\Omega|^{2}\Omega\Delta n\mu_{ab}}{4\Delta^{4}}=\frac{2N|\mu_{ab}|^{4}}{\hbar^{3}\Delta^{3}}|E|^{2}E \tag{3.42}
\end{equation}
\begin{equation}
    \chi^{(3)}=\frac{2N|\mu_{ab}|^{4}}{3\epsilon_{0}\hbar^{3}\Delta^{3}} \tag{3.43}
\end{equation}
