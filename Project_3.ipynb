## Aim

To develop methods for predicting the intensity of reflected harmonic waves generated at material interfaces in nonlinear optical processes.

## Motivation

Studying second harmonic wave generation at the surface of materials with distinct optical properties is scientifically motivated by its ability to probe surface and interface properties, providing valuable insights into structural and electronic characteristics. This research has practical implications for materials engineering, surface characterization, biomedical imaging, quantum optics, and fundamental physics, making it a multidisciplinary field with the potential for significant scientific discoveries and technological advancements.

The transmitted wave gives rise to polarisation given by:<br>

\mathbf{k}_s(\omega_s) = 2\mathbf{k}_T(\omega_i) \quad \mathbf{\hat{p}} = \mathbf{p} / |\mathbf{\hat{p}}|
$$

 Here we are considering the case of polarization $\textbf{p}$ perpendicular to the plane of incidence

## Nonlinear Polarization Radiation

Non linear polarisation gives rise reflected wave at $ 2\omega_i$ which is governed by: 
<br><br>
$$
\nabla^2 \mathbf{E}(\omega_s) + [\epsilon (\omega_s) \omega_s^2 / c^2] \mathbf{E} (\omega_s) = -(\omega_s^2 / \epsilon_0 c^2) \mathbf{p}_\perp e^{i \mathbf{k_s} \cdot \mathbf{r} } \tag{1}
$$
<br>
Here  $\epsilon(w)$ is taken to be $\epsilon_T(w)$ in the nonlinear medium and $\epsilon_R(w)$ in the linear medium
<br><br>
For the homogenous equation: $\hspace{0.3cm} \nabla^2 \mathbf{E}(\omega_s) + [\epsilon (\omega_s) \omega_s^2 / c^2] \mathbf{E} (\omega_s)=0 $ <br>
the soultion is $\mathbf{E}_T(\omega_s) = \mathbf{A}_T(\omega_s)e^{i\mathbf{k}_T(\omega_s) \cdot \mathbf{r}} $
<br><br>
Let the particular solution to be $\mathbf{E}_T(\omega_s) = k\mathbf{p}_\perp e^{i \mathbf{k_s} \cdot \mathbf{r}} 
$ Use the relation: $\nabla e^{i\textbf{k}\cdot\textbf{r}}= -i.k.e^{i\textbf{k}\cdot\textbf{r}}$
<br> On subsituting the above solution in equation 1, we get:
<br> $$ k\textbf{p}_{\perp}e^{i\textbf{k}_s\cdot\textbf{r}}[-k^2_s + k^2_T] = -\frac{\omega^2_s}{\epsilon_0c^2}\textbf{p}_{\perp}e^{i\textbf{k}_s\cdot\textbf{r}}$$
 And $ k=\frac{(\omega_s^2 / \epsilon_0^2)}{|k_s|^2 - |k_T (\omega_s)|^2}$
 <br> <br>So we get the solution to be:
 $$ \mathbf{E}_T(\omega_s) = \mathbf{A}_T(\omega_s)e^{i\mathbf{k}_T(\omega_s) \cdot \mathbf{r}} + \frac{(\omega_s^2 / \epsilon_0^2)}{|k_s|^2 - |k_T (\omega_s)|^2} \mathbf{p}_\perp e^{i \mathbf{k_s} \cdot \mathbf{r} } \tag{2} $$
 <br><br>
 where $k_s = \ \epsilon^{1/2}(\omega_s)  \omega_s / c \quad$ and  $\quad k_T(\omega_s) = \epsilon^{1/2}_T (\omega_s) \omega_s / c$
## Boundary Conditions



<br><br>
The boundary conditions requires for that the components of $\textbf{E}$ and $\textbf{H}$ tangential to the plane of interface be continuous. From thsese conditions to be satisfied , the exisitence of second-harmonic reflected wave is required. The reflected wave is given by:
<br>
$$ \mathbf{E}_R(\omega_s) = \mathbf{A}_R(\omega_s)e^{i\mathbf{k}_R(\omega_s) \cdot \mathbf{r}} \tag{3}$$<br><br>

For the boundary conditions to satisfy at each point of interface the wave-vectors $\textbf{k}_s$, $\textbf{k}_T(\omega_s)$ and $\textbf{k}_R(\omega_s)$ should have identical compoents along the interface, i.e
<br>
$$ k_x^s = k_x^R(\omega_s) = k_x^T(\omega_s) \tag{4} $$
<br> Here $ k_s = 2k_T(\omega_i) \tag{5}$
<br><br>
The magnitudes of wavevectors can be expressed in terms of dielectric constants:
<br>
$$
k_T(\omega_s) = \epsilon_T^{1/2}(\omega_s) \omega_s /c \tag{6}
$$
$$
k_R(\omega_s) = \epsilon_R^{1/2}(\omega_s) \omega_s /c \tag{7}
$$
$$
k_i(\omega_i) = \epsilon_R^{1/2}(\omega_i) \omega_i /c \tag{8}
$$
<br>
Similarly we can write $k_s$ interms of the dubious dielectric constant $\epsilon_s$:
$$
k_s = \epsilon_s^{1/2} \omega_s /c \tag{9}
$$
### Nonlinear optical generalization of Snell's law

From the figure 1 we can write the x components of the wave vectors:
<br> $$k^s_x= k_ssin(\theta_s)= (\epsilon^{1/2}_s\omega_s/c )sin(\theta_s)$$
$$k^T_x(\omega_S)= k_T(\omega_s)sin(\theta_T)= (\epsilon^{1/2}_T(\omega_s)\omega_s/c )sin(\theta_T)$$
$$k^R_x(\omega_s)= k_R(\omega_s)sin(\theta_R)= (\epsilon^{1/2}_R(\omega_s)\omega_s/c) sin(\theta_R)$$
<br>
so we get: $ \epsilon_R^{1/2}(\omega_s)sin \theta_R = \epsilon_T^{1/2}(\omega_s)sin \theta_T = \epsilon_s^{1/2} sin \theta_s$ (from equation 4)
<br>
And also $k^i_x= (\epsilon_R^{1/2}(\omega_i)\omega_i/c)sin(\theta_i)$ and $k^R_x= 2k^i_x$ <br><br>
Finally we get the relation:
<br>
$$
\epsilon_R^{1/2}(\omega_i)sin \theta_i = \epsilon_R^{1/2}(\omega_s)sin \theta_R = \epsilon_T^{1/2}(\omega_s)sin \theta_T = \epsilon_s^{1/2} sin \theta_s \tag{10}
$$

<br><br>
The $\textbf{p}_{\perp}$ is in the y direction, this will generate the electric field in the y direction and the magnetic fields in the x-z plane<br>
We are going to equate the amplitude of reflected fileds with the transmitted and polarized fileds of second-harmonic.
<br>
The continuity of the tangential components of $\textbf{E}$ gives:
<br>
$$E_{y} : A^R _ \perp (\omega_{s}) = A^T_{\perp} (\omega_{s}) + \frac{p_{\perp}}{ \epsilon _0  [\epsilon _s -\epsilon _T (\omega _s)]} \tag{10} $$
<br><br>
The continuity of the tangential components of $\textbf{H}$ gives:
<br>
$$ H_{x} : {-\epsilon_R^{1/2}} (\omega _s) {A^R _{\perp}(\omega{s})}  cos \theta _R = {\epsilon^{1/2}_T} (\omega _s) {A^T _{\perp}(\omega_s)} cos {\theta _T} + \frac{p {\perp} cos{\theta _s}{\epsilon ^{1/2} _s}} { \epsilon _0  [\epsilon _s -\epsilon _T (\omega _s)] } \tag{11}$$
<br><br>





## Simplifying further

On solving the equations (10) and (11) we get:
<br>
$$
A_{\perp}^{R} = \frac{p_{\perp}}{\epsilon_0[\epsilon_s-\epsilon_T(\omega_s)]}[\frac{\epsilon_T^{1/2}(\omega_s)cos(\theta_T)-\epsilon_s^{1/2}(\omega_s)cos(\theta_s)}{\epsilon_R^{1/2}(\omega_s)cos(\theta_R) + \epsilon_T^{1/2}(\omega_s)cos(\theta_T) }]
$$
$$
A_{\perp}^{T} = -\frac{p_{\perp}}{\epsilon_0[\epsilon_s-\epsilon_T(\omega_s)]}[\frac{\epsilon_s^{1/2}(\omega_s)cos(\theta_s)+\epsilon_R^{1/2}(\omega_s)cos(\theta_R)}{\epsilon_R^{1/2}(\omega_s)cos(\theta_R) + \epsilon_T^{1/2}(\omega_s)cos(\theta_T) }]
$$

On subsituting the values of $A^R_{\perp}$ and $A^T_{\perp}$ we get:
<br><br>
$$ 
E^R_{\perp}(\omega_{s}) = \frac{ -p_{\perp}e^{i \textbf{k}_R(\omega_s) \cdot \textbf{r}}}{\epsilon_0[\sqrt{\epsilon_T}(\omega_s) \cos{\theta_T} + \sqrt{\epsilon_R}(\omega_s) \cos{\theta_R}][\sqrt{\epsilon_T}(\omega_s) \cos{\theta_T} + \sqrt{\epsilon_s} \cos{\theta_s}]}
$$

$$ \equiv A^R_{\perp}(\omega_{s}) e^{i k_R(\omega_s) \cdot r}$$
<br>
<br>
$$
E^T_{\perp}(\omega_{s}) = \frac{- p_{\perp}}{\epsilon_0[\epsilon_T(\omega_s) -  \epsilon_s]} \left[ e^{i \textbf{k}_s \cdot \textbf{r}} - \frac{\epsilon_s^{1/2} \cos(\theta_s) + \epsilon_R^{1/2}(\omega_s) \cos(\theta_R)}{\epsilon_T^{1/2}(\omega_s) \cos(\theta_T) + \epsilon_R^{1/2}(\omega_s) \cos(\theta_R)} e^{i \textbf{k}_T(\omega_s) \cdot \textbf{r}} \right] \tag{12}
$$
<br>
Now we look at $\mathbf{k}_s - \mathbf{k}_T$
<br>
$$ \textbf{k}_s - \textbf{k}_T={\small \triangle}
 k \hat{z} = \frac{\omega_s}{ c}[ \epsilon^{1/2}_s cos{\theta_s} - \epsilon^{1/2}_T (\omega _s) cos{\theta_T} ] \hat{z} $$
<br>
This lie along z axis , we put this result in the equation of $E^T_{\perp}(\omega_s)$<br>
$$E^T_{\perp}(\omega_{s})=[A^R_{\perp}(\omega_s) + {\frac{{}(\omega^2_s /\epsilon_0 c^2) p_{\perp}}{2 {k_T}(\omega_s)}}{\left( \frac{e^{i\triangle kz} -1}{\triangle k} \right)}] e^{i\textbf{k}_{T}(\omega_s)\cdot\textbf{r}} \tag{13}$$ <br>

$$
\equiv A^T_{\perp}(\omega_s) e^{i{\textbf{k}_T}(\omega_s)\cdot\textbf{r}}$$
The amplitude is spatially varying due to imperfect phase matching .
Now , assuming $\triangle kz < 1$ we get $A^T_{\perp}(\omega_s)$ in terms of $A^R_{\perp}(\omega_s)$
$$\mathrm{A}^{T}{\perp}(\omega_s) = \mathrm{A}^{R}_{\perp}(\omega_s) + \frac{(\omega /c)^2 p_\perp (iz)}{2\epsilon_0k_{T} (\omega_s)} =       \mathrm{A}^{R}_{\perp}(\omega_s)     +
\frac{i(\omega /c) p_{\perp}z}{2\epsilon_0 \epsilon^{1/2} (\omega_s)} \tag{14}$$
<br>
We also see from equation 12 that the order of magnitude of  $\mathrm{A}^{R}_{\perp}(\omega_s)$ is given by 
$$
{A}^{R}_{\perp} (\omega_s) \hspace{0.3cm}\widetilde{-} \hspace{0.1cm}-\frac{ p_{\perp}}{4\epsilon_0\epsilon} \tag{15}
$$
We use equation 14 and
$4\epsilon =  \sqrt{\epsilon_T}(\omega_s) \cos{\theta_T} + \sqrt{\epsilon_R}(\omega_s) \cos{\theta_R}][\sqrt{\epsilon_T}(\omega_s) \cos{\theta_T} + \sqrt{\epsilon_s} \cos{\theta_s}
$ to get
$$
\mathrm{A}^{T}_{\perp} (\omega_s) \hspace{0.3cm}\widetilde{-} \hspace{0.1cm}-\frac{\pi p_{\perp}}{4\epsilon_0\epsilon} \left[1 - 2ik_{T} (\omega_s) z\right] \tag{16}
$$

The surface term makes contribution comparable to bulk term for thickness t :
$$
t = \frac{\lambda}{4\pi}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
def eps_r(omega):
  return 1

def eps_t(omega):
  return 12.35

def xi2(omega, z):
  xi = np.zeros(len(z))
  t = 5.2e-11

  for j in range(len(z)):
    if (z[j] < t):
      xi[j] = 150e-12
    if (z[j] >= t):
      xi[j] = 150e-12*np.exp(-(z[j] - t)/t)**2

  return xi

def E(z, A, omega_i, theta_i):

  omega_s = 2*omega_i
  epsilon_r = eps_r(omega_i)
  epsilon_t = eps_t(omega_s)
  xi_2i = xi2(omega_i, z)

  c = 3e8
  epsilon_0 = 8.85e-12
  mu_0 = 4*np.pi*1e-7

  k_i = (epsilon_r**0.5)*omega_i/c
  k_r = (epsilon_r**0.5)*omega_s/c
  k_t = (epsilon_t**0.5)*omega_s/c
  k_s = 2*k_t
  delta_k = k_s - k_t
  epsilon_s = (k_s*c/omega_s)**2

  theta_r = theta_i
  theta_t = np.arcsin((epsilon_r**0.5)*(np.sin(theta_i))/(epsilon_t**0.5))
  theta_s = np.arcsin((epsilon_r**0.5)*(np.sin(theta_i))/(epsilon_s**0.5))

  p = epsilon_0*xi_2i*A*A
  p_perp = p

  t1 = -1*mu_0*p_perp
  t2 = (epsilon_t**0.5)*np.cos(theta_t)+(epsilon_r**0.5)*np.cos(theta_r)
  t3 = (epsilon_t**0.5)*np.cos(theta_t)+(epsilon_s**0.5)*np.cos(theta_s)
  A_perp_R = (t1/(t2*t3))
  E_perp_R = A_perp_R*np.exp(k_r*z*1j)

  t4 = (mu_0*(omega_s/c)*(omega_s/c)*p_perp)/(2*k_t)
  t5 = (np.exp(delta_k*z*1j))/(delta_k)
  A_perp_t = (A_perp_R + t4*t5)
  E_perp_T = A_perp_t*np.exp(k_t*z*1j)

  return E_perp_R, E_perp_T
```

```python
#PLOTTING INCIDENT WAVE VS PROPOGATION ANGLE

t = 100e-11

z0 = np.linspace(0, -t, 10000)
z = np.linspace(0, t, 10000)

c = 3e8
A = 5e10
omega_i = 5.6e14
k_i = (eps_r(omega_i)**0.5)*omega_i/c
y0 = A*np.exp(k_i*z*1j)
y1, y2 = E(z, A, 5.6e14, np.pi/5)

#plt.plot(z0, np.real(y0), color = 'red')
plt.plot(z0, np.real(y1), color = 'blue')
plt.plot(z, np.real(y2), color = 'green')

plt.xlabel("z")
plt.ylabel("E")
plt.title("E vs z")

plt.show()
```

```python
def A_RT(z, A, omega_i):

  omega_s = 2*omega_i
  c = 3e8
  epsilon_0 = 8.85e-12
  epsilon = 1
  epsilon_t = eps_t(omega_s)
  k_t = (epsilon_t**0.5)*omega_s/c

  xi_2i = xi2(omega_i, z)

  p_perp = epsilon_0*xi_2i*A*A

  A_perp_R=(-1*p_perp)/(4*epsilon_0*c*c*epsilon_0)
  A_perp_R = A_perp_R*np.ones(len(z))

  m1=np.pi*p_perp
  m2=1-2*k_t*z*1j
  m3=4*epsilon_0*c*c*epsilon
  A_perp_T=m1*m2/m3

  return A_perp_R, A_perp_T
```

# Derivation of the Wave Equation for Nonlinear Polarization



We begin with Maxwell's equations:

**Gauss's Law for Electric Fields:**
$$\nabla \cdot \tilde{\mathbf{D}} = \tilde{\rho} \tag{1}$$

**Gauss's Law for Magnetic Fields:**
$$\nabla \cdot \tilde{\mathbf{B}} = 0 \tag{2}$$

**Faraday's Law:**
$$\nabla \times \tilde{\mathbf{E}} = -\frac{\partial \tilde{\mathbf{B}}}{\partial t} \tag{3}$$

**Ampère's Law:**
$$\nabla \times \tilde{\mathbf{H}} = \frac{\partial \tilde{\mathbf{D}}}{\partial t} + \tilde{\mathbf{J}} \tag{4}$$

In a region devoid of free charges and currents ($\tilde{\rho} = 0$, $\tilde{\mathbf{J}} = 0$) and assuming nonmagnetic materials ($\tilde{\mathbf{B}} = \mu_0 \tilde{\mathbf{H}}$), the nonlinear polarization $\tilde{\mathbf{D}}$ is given by:

$$\tilde{\mathbf{D}} = \epsilon_0 \tilde{\mathbf{E}} + \tilde{\mathbf{P}} \tag{5}$$

Taking the curl of Faraday's Law (Eq. $3$), we obtain:

$$\nabla \times \nabla \times \tilde{\mathbf{E}} + \mu_0 \frac{\partial^2}{\partial t^2} \tilde{\mathbf{D}} = 0 \tag{6}$$

Using Eq. $5$, we eliminate $\tilde{\mathbf{D}}$ and arrive at the general form of the wave equation in nonlinear optics:

$$\nabla \times \nabla \times \tilde{\mathbf{E}} + \frac{1}{c^2} \frac{\partial^2}{\partial t^2} \tilde{\mathbf{E}} = -\frac{1}{\epsilon_0 c^2} \frac{\partial^2 \tilde{\mathbf{P}}}{\partial t^2} \tag{7}$$

Here, $\tilde{\mathbf{P}}$ can be split into its linear and nonlinear components ($\tilde{\mathbf{P}} = \tilde{\mathbf{P}}^{(1)} + \tilde{\mathbf{P}}^{\mathrm{NL}}$). Defining $\tilde{\mathbf{D}}^{(1)} = \epsilon_0 \tilde{\mathbf{E}} + \tilde{\mathbf{P}}^{(1)}$, the wave equation simplifies to:

$$\nabla^2 \tilde{\mathbf{E}} - \frac{1}{\epsilon_0 c^2} \frac{\partial^2 \tilde{\mathbf{D}}^{(1)}}{\partial t^2} = \frac{1}{\epsilon_0 c^2} \frac{\partial^2 \tilde{\mathbf{P}}^{\mathrm{NL}}}{\partial t^2} \tag{8}$$

In the case of a lossless, dispersionless medium, $\tilde{\mathbf{D}}^{(1)} = \epsilon_0 \epsilon^{(1)} \cdot \tilde{\mathbf{E}}$, where $\epsilon^{(1)}$ is a frequency-independent dielectric constant. For this scenario, the wave equation takes the form:

$$\nabla^2 \tilde{\mathbf{E}} - \frac{\epsilon^{(1)}}{c^2} \frac{\partial^2 \tilde{\mathbf{E}}}{\partial t^2} = \frac{1}{\epsilon_0 c^2} \frac{\partial^2 \tilde{\mathbf{P}}^{\mathrm{NL}}}{\partial t^2} \tag{9}$$

In the presence of dispersion, each frequency component of the field is considered separately. The wave equation for each component becomes:

$$\nabla^2 \tilde{\mathbf{E}}_n - \frac{\boldsymbol{\epsilon}^{(1)}(\omega_n)}{c^2} \cdot \frac{\partial^2 \tilde{\mathbf{E}}_n}{\partial t^2} = \frac{1}{\epsilon_0 c^2} \frac{\partial^2 \tilde{\mathbf{P}}_n^{\mathrm{NL}}}{\partial t^2} \tag{10}$$

For a dissipative medium, the dielectric tensor is complex:

$$\mathbf{D}_n^{(1)}(\mathbf{r}) = \epsilon_0 \boldsymbol{\epsilon}^{(1)}(\omega_n) \cdot \mathbf{E}_n(\mathbf{r}) \tag{11}$$

This leads to the Helmholtz equation in the frequency domain:

$$\nabla^2 \mathbf{E}_n(\mathbf{r}) + \frac{\omega_n^2}{c^2} \boldsymbol{\epsilon}^{(1)}(\omega_n) \cdot \mathbf{E}_n(\mathbf{r}) = -\frac{\omega_n^2}{\epsilon_0 c^2} \mathbf{P}_n^{\mathrm{NL}}(\mathbf{r}) \tag{12}$$
